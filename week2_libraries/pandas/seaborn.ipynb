```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# seaborn: seaborn is a data visualization library built on top of matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
# common seaborn plots
# heatmap
# pairplot
# displot
# kdeplot
# boxplot
# violinplot
# stripplot
# swarmplot
# countplot
# barplot
# lineplot
# scatterplot
# histplot
```

```python
# heatmap
data=np.random.rand(10,10)
sns.heatmap(data)
plt.title("Heatmap")
plt.show()
```

```python
# heatmap
data=np.random.rand(10,10)
sns.heatmap(data,annot=True, cmap='YlGnBu') #cmap: cmap will color the heatmap example: YLGnBu means yellow green blue, coolwarm means cool warm
# annot means annotation of the heatmap values
plt.title('Heatmap')
plt.show()
```

```python
# pairplot
data=np.random.randn(100,2)

df=pd.DataFrame(data,columns=["x","y"])
sns.pairplot(df,hue="x")
plt.title("Pairplot")
plt.show()
```

```python
# create a basic plots with mathplotlib

year=[2001,2002,2003,2004,2005]
sales=[200,300,400,500,600]
plt.plot(year,sales,label='sales', color='red', marker='o', linestyle='--')
plt.title('Sales Data')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
plt.scatter(year, sales, label="Sales", color="red", linewidths=2)
plt.title("Sales")
plt.xlabel("Year")
plt.ylabel("Sales") 
plt.legend()
plt.show()
```

```python
# bar chart
categorical=['electronics', 'furniture', 'clothing']
revenue=[250,200,150]
plt.bar(categorical,revenue,color='green')
plt.title('Revenue by category')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(df)


del df['species'] #here we delete species column because it is not a numeric value and it is string 

# calculate correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# creata a histogram with multiple datasets overlaid



df=sns.load_dataset("iris")
sns.histplot(data=df,x="sepal_length",kde=True,hue="species")
plt.show()
```
