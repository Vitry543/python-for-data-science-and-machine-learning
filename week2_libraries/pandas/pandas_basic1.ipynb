#data cleaning and preparation with pandas

```python
# handling missing data: missing values are denoted by NaN and are not included in the calculations.


# methods to handle missing data:
# 1. dropna(): drops the rows or columns with missing values #data missing values
# 2. fillna(): fills the missing values with a specified value #fill missing values
# 3. interpolate(): interpolates the missing values #interpolate missing values
# 4. isnull(): checks if a value is missing #check missing values
# 5. notnull(): checks if a value is not missing #check not missing values
# 6. isna(): checks if a value is missing #check missing values
# 7. notna(): checks if a value is not missing #check not missing values
```

```python
# examples
df=df.dropna()
# dropna() will drop the rows with missing values
df=df.dropna(how='all') # dropna(how='all') will drop the rows with all missing values
df=df.dropna(axis=1) # dropna(axis=1) will drop the columns with missing values

# fillna() will fill the missing values
df=df.fillna()
df=df.fillna(method='ffill') #ffill:forward fill # fillna(method='ffill') will fill the missing values with the previous value
df=df.fillna(method='bfill') #bfill:backward fill # fillna(method='bfill') will fill the missing values with the next value
df=df.fillna(method='pad') #pad:forward fill # fillna(method='pad') will fill the missing values with the previous value
df=df.fillna(method='backfill') #backfill:backward fill # fillna(method='backfill') will fill the missing values with the next value

df=df.fillna(0) # fillna(0) will fill the missing values with 0
df=df.fillna(method='ffill',limit=1) # fillna(method='ffill',limit=1) will fill the missing values with the previous value and limit=1 will fill the missing values with the previous value only once

df=df.fillna(axis=1) # fillna(axis=1) will fill the missing values with the next value

df['column_name']=df['column_name'].fillna(0) # fillna(0) will fill the missing values with 0

```

```python
# interpolation

df.interpolate(method="linear") # linear interpolation : it will interpolate the missing values linearly
df.interpolate(method="quadratic") # quadratic interpolation : it will interpolate the missing values quadratically

df['colum_name']=df['colum_name'].interpolate(method="linear") # it will interpolate the missing values linearly in the column
```

```python
# combining and merging dataframes
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(df)
# concatenating dataframes
df1=pd.concat([df,df],axis=0)
print(df1)
df1=pd.concat([df,df],axis=1)
print(df1)

# merging dataframes
df2=pd.merge(df,df,on="species")  #here on="species" means that we are merging on the basis of species
print(df2)

df2=pd.merge(df,df,on="species",how="inner") #how="inner" means that we are merging on the basis of inner join
print(df2) #inner join means that we are merging on the basis of common values

df2=pd.merge(df,df,on="species",how="outer") #how="outer" means that we are merging on the basis of outer join
print(df2) #outer join means that we are merging on the basis of both dataframes

df2=pd.merge(df,df,on="species",how="left") #how="left" means that we are merging on the basis of left join
print(df2) #left join means that we are merging on the basis of left dataframe

df2=pd.merge(df,df,on="species",how="right") #how="right" means that we are merging on the basis of right join
print(df2) #right join means that we are merging on the basis of right dataframe

```
