```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplotlib : matplotlib is a python library for data visualization. It is a low level library for creating static, animated, and interactive visualizations in python.
# seaborn : seaborn is a python library for data visualization. It is a high level library for creating static, animated, and interactive visualizations in python.

# differnece between matplotlib and seaborn : 
# 1.mathplot for General purpose plotting, high customization, whereas seaborn is for statistical data visualization.
# 2.mathplot Requires more code for complex/attractive plots (steep learning curve, more time consuming), whereas seaborn is for simple plots (easy to use, less time consuming).
```

```python
# basic plot
x=[1,2,3,4,5]
y=[2,3,5,7,11]
plt.plot(x,y)
plt.show()
```

```python
# basic plots
# lineplot : lineplot()
# scatterplot : scatterplot()
# histogram : histogram()
# boxplot : boxplot()
# violinplot : violinplot()
# barplot : barplot()
# heatmap : heatmap()
```

```python
# line plot
plt.plot([1,2,3,4],[1,4,9,16],label="trend",color="orange",linestyle="-",marker="o")
plt.title("Line plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["lineplot"])
plt.show()
```

```python
# line plot
plt.plot([1,2,3,4],[1,4,9,16],label="trend",color="orange",linestyle="--",marker="o")
plt.title("Line plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["lineplot"])
plt.show()
```

```python
# scatter plot
plt.scatter(x, y,color="green")
plt.title("Scatter Plot",color="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["dataser1"])
plt.show()
```

```python
# bar chart
categorical=["a",'b','v']
values=[10,20,30]
plt.bar(categorical,values)
plt.show()
```

```python
# histogram
import matplotlib.pyplot as plt
data=[1,2,3,3,4,5,5,4,3,2,1]
plt.hist(data)
plt.show()
```

```python
# histogram
import matplotlib.pyplot as plt
data=[1,2,2,3,3,3,4,4,4,4]
plt.hist(data,bins=4, color='green', edgecolor='black') #bins = number of bins. bins means number of intervals between min and max
plt.title('Histogram')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()
```
