```python
# grouping data by category
# why group data: to perform analysis on data and get insights of data

# groupby in pandas
# groupby is a function that groups data by a column
```

```python
import pandas as pd
df=pd.read_csv("iris.csv")
print(df)
```

```python
grouped=df.groupby("species")
print(grouped)
grouped.mean()
```

```python
# operations
# 1.iterate over groups
# 2.apply aggregate functions
```

```python
for name, group in grouped:
    print("Group name:", name)
    print("Group:", group)

```

```python
# apply aggregate function
aggregated = grouped.agg({"sepal_length": "mean", "sepal_width": "mean"})
print(aggregated)
print(grouped.mean())
print(grouped.agg({"sepal_length": "mean", "sepal_width": "mean"}))
print(grouped.agg({"sepal_length": "mean", "sepal_width": "mean"}).mean())
```

```python
# aggregation funciton: Using groupby
df.groupby("species")["petal_length"].mean()
```

```python
df.groupby("species")["petal_width"].mean()
```

```python
df.groupby("species").agg({"sepal_length": ["mean",'max','min']})
```

```python
# using pivot table
# pivot table is a way to summarize data and it is a way to transform data
# pivot_table will create a new table with the index as the rows and the columns as the columns

# aggregate function: mean, sum, max, min, count, std, var, median

# synatx of pivot_table: df.pivot_table(values="numeric_column_name",index="column_name",columns="numeric_column_name",aggfunc="function_name")

pivot=df.pivot_table(values="petal_length",index="species",columns="petal_width",aggfunc="mean")
# in the above code, values is the column we want to aggregate, index is the column we want to group by, columns is the column we want to group by, aggfunc is the function we want to use to aggregate
pivot
```

```python
# custom aggeration function
def range_fuc(x):
    return x.max() - x.min()

iris = pd.read_csv("iris.csv")
iris.groupby("species").agg({"sepal_length": range_fuc})
```

```python
def range_fuc(x):
    return x.max() - x.min()

iris = pd.read_csv("iris.csv")
iris.groupby("species")["sepal_length"].agg(range_fuc)
```

```python
# calculating summary statistics for grouped data
# common statistics:
# count, mean, std, min, 25%, 50%, 75%, max

# synatx iris.groupby("column_name").statistic()

# statistics for species
print(iris.groupby("species").mean())
print(iris.groupby("species").std())
print(iris.groupby("species").min())
print(iris.groupby("species").max())
print(iris.groupby("species").count())
print(iris.groupby("species").median())
print(iris.groupby("species").var())
print(iris.groupby("species").sem())
```

```python
# statistics for each species
# synatx iris.groupby("column_name")["numeric_column_name"].function()

print(iris.groupby("species")["sepal_length"].mean())
print(iris.groupby("species")["sepal_length"].std())
print(iris.groupby("species")["sepal_length"].min())
print(iris.groupby("species")["sepal_length"].max())
print(iris.groupby("species")["sepal_length"].count())
print(iris.groupby("species")["sepal_length"].median())
print(iris.groupby("species")["sepal_length"].var())
print(iris.groupby("species")["sepal_length"].sem())
```

```python
# multi-aggregation
# syntax iris.groupby('column_name').agg({'column_name': 'aggregation_function'})
iris = pd.read_csv('iris.csv')
iris.groupby('species').agg({'sepal_length': ['mean', 'std'], 'sepal_width': 'mean', 'petal_length': 'mean', 'petal_width': 'mean'})
```

```python
# create a dataset of sales data and group it by month and calculate the total sales for each month
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# create a dataset of sales data and group it by month and calculate the total sales for each month
sales_data = pd.read_csv('sales_data.csv')
sales_data['month'] = pd.to_datetime(sales_data['date']).dt.month
sales_data.groupby('month').sum()
sales_data.groupby('month').sum().plot(kind='bar')
```

```python
# create a dataset of sales data and group it by region or product type

import pandas as pd
import numpy as np
sales_data = [
    {"id": 1, "product": "Laptop", "region": "North", "amount": 1200},
    {"id": 2, "product": "Mouse", "region": "South", "amount": 25},
    {"id": 3, "product": "Keyboard", "region": "North", "amount": 75},
    {"id": 4, "product": "Laptop", "region": "East", "amount": 1300},
    {"id": 5, "product": "Mouse", "region": "North", "amount": 20},
    {"id": 6, "product": "Monitor", "region": "West", "amount": 300},
    {"id": 7, "product": "Laptop", "region": "South", "amount": 1150},
    {"id": 8, "product": "Keyboard", "region": "West", "amount": 80},
    {"id": 9, "product": "Monitor", "region": "North", "amount": 320},
    {"id": 10, "product": "Mouse", "region": "East", "amount": 30}
]
sales_df = pd.DataFrame(sales_data)
sales_by_region=sales_df.groupby('region').agg({'amount': 'sum'})
sales_by_product=sales_df.groupby('product').agg({'amount': 'sum'})
print(sales_by_region)
print(sales_by_product)
```

```python
# use pivot_table to calculate the total sales by region and product
sales_df.pivot_table(index='region', columns='product', values='amount', aggfunc='sum')
```

```python
# create a custom aggregation function to calculate the variance of each group

sales_data = [
    {"id": 1, "product": "Laptop", "region": "North", "amount": 1200},
    {"id": 2, "product": "Mouse", "region": "South", "amount": 25},
    {"id": 3, "product": "Keyboard", "region": "North", "amount": 75},
    {"id": 4, "product": "Laptop", "region": "East", "amount": 1300},
    {"id": 5, "product": "Mouse", "region": "North", "amount": 20},
    {"id": 6, "product": "Monitor", "region": "West", "amount": 300},
    {"id": 7, "product": "Laptop", "region": "South", "amount": 1150},
    {"id": 8, "product": "Keyboard", "region": "West", "amount": 80},
    {"id": 9, "product": "Monitor", "region": "North", "amount": 320},
    {"id": 10, "product": "Mouse", "region": "East", "amount": 30}
]
sales_df = pd.DataFrame(sales_data)

def variance(x):
    return np.var(x)

sales_df.groupby('region').agg({'amount': variance})
```
