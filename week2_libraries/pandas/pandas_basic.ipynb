```python
# # pandas is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.
# pandas in ai is used for data analysis and data science.
# pandas is a package for scientific computing with Python.
# pandas types:
# 1. series
# 2. dataframe
```

```python
# pandas series: is a one-dimensional array-like object containing a sequence of values and an associated array of data labels, called its index.
# pandas dataframe: is a two-dimensional, size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns).
```

```python
s=pd.Series([1,2,3,4,5])
print(s)
s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
print(s)

s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"],name="numbers")
print(s)

s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"],name="numbers",dtype="float")
print(s)

s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"],name="numbers",dtype="float",copy=True)
print(s)

s=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"],name="numbers",dtype="bool",copy=True,)
print(s)
```

```python
df=pd.read_csv('titanic.csv')

# print(df)
# operations
# for viewing data

print(df.head()) #it will print first 5 row of a paricular dataset
print(df.tail()) #it will print last 5 row of a paricular dataset

print(df.tail(3))

print(df.info())
print(df.describe())

# selecting and indexing the data
print(df["Name"])
print(df[['Name','Age']])

# filtering the data
print(df[df['Age']>25])

# selecting by particular postion
print(df.iloc(0))
print(df.iloc[: ,0])

# selecting by label
print(df.loc[0])
print(df.loc[: , "Name"])
```

```python
# select specific columns and filter row
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
seleceted_columns=df[["species","sepal_length"]]
print("selected columns: \n",seleceted_columns)

# filter row
filtered_row=df[df["species"]=='setosa']
print("filtered row: \n",filtered_row)

filtered_row=df[(df["species"]=='setosa') & (df["sepal_length"]>5)]
print("filtered row: \n",filtered_row)
```
