```python
# loading the data
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)

# taks1
# inspect the data
print(df.info())
print()
print(df.describe())

# handle missing values
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])

# remove duplicates
df=df.drop_duplicates()

# filter data: Passagers in first class 
first_class=df[df["Pclass"]==1]
print("first class passagers:\n",first_class.head())

print('\n')

# task2
# BarChart : survival rate by class
survival_by_class=df.groupby("Pclass")["Survived"].mean()
survival_by_class.plot(kind="bar",color='skyBlue')
plt.title("Survival Rate by class")
plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.show()

# histogram : age distribution
sns.histplot(df['Age'],kde=True,bins=20,color='purple')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# task3
# scatter plot : fare vs age
# sns.scatterplot(x='Age',y='Fare',data=df)
plt.scatter(df['Age'],df['Fare'],color='red',alpha=0.5)
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()


# task4

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url="https://raw.githubusercontent.com/datamole-ai/edvart/refs/heads/main/example-datasets/pollution.csv"
data=pd.read_csv(url)
print(data)
# anaylse the data
print(data.info())
print(data.describe())

# check for missing values
print("\nMissing values:")
print(data.isnull().sum())

# handling missing values
data['pollution']=data['pollution'].fillna(data['pollution'].mean())

# check for duplicate values
print("\nDuplicate values:")
print(data.duplicated().sum())



# visualizing the data
# barchart: pollution levels by month

pollution_by_dew=data.groupby('dew')['pollution'].mean().reset_index()
sns.barplot(x='dew',y='pollution',data=pollution_by_dew)
plt.title('Pollution levels by dew')
plt.show()




```
