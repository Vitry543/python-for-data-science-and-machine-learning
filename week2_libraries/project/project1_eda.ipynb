```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url="https://raw.githubusercontent.com/datamole-ai/edvart/refs/heads/main/example-datasets/auto.csv"
data=pd.read_csv(url)
print(data)
# anaylse the data
print(data.info())
print(data.describe())
print(data.isnull().sum())

# handling missing values
data['mpg']=data['mpg'].fillna(data['mpg'].mean())

# check for duplicate values
print("\nDuplicate values:")
print(data.duplicated().sum())

# visualizing the data
# barchart: mpg by origin

mpg_by_origin=data.groupby('origin')['mpg'].mean()
mpg_by_origin.plot(kind='bar',color='skyblue')
plt.title('MPG by Origin')
plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()

# histogram: mpg
sns.histplot(data['mpg'],kde=True,color='skyblue')
plt.title('MPG Distribution')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

# scatter plot: mpg vs weight
plt.scatter(data['mpg'],data['weight'],color='skyblue')
# sns.regplot(x='mpg',y='weight',data=data,color='skyblue')
plt.title('MPG vs Weight')
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.show()

plt.scatter(data['displacement'],data['acceleration'],color='skyblue')
plt.title('Displacement vs Acceleration')
plt.xlabel('Displacement')
plt.ylabel('Acceleration')
plt.show()

# create a dashboard for your findings using plotly or dash
# dashboard for displacement vs acceleration
# plotly express
import plotly as 


```
