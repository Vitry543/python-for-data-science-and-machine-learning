```python
# Deep Learning:
    # -Subset of ML that uses Artifical Neural network(ANNs) with multiple layers (deep Architectures) to model and learn complex patterns in data

    # Key Features:
        # -Automatically extracts relevant features from raw data, eliminating the need for manual feature engineering

# ML VS Deep Learning

# Features                             Machine learning                                     Deep Learning
# 1.Feature engineering               Manual Feature selection                            Automatic feature extraction
# 2.Algorithms                        Linear regression, SVM, etc..                       Neural Network
# 3.Data Dependency                   Works well with small datasets                      Requires large dataset
# 4.Computation                       Computationally less intensive                      Requires GPUs/ TPUs for training
```

```python
# Overview of Artificial Neural Network(ANNs):

# Structure of a Neural Network:
    # Input layer:
        # -Accepts input data features
    # Hidden layers:
        # Performs computations to extract patterns
    # output layer:
        # Produces predictions or classifications

# Key Components:
    # Neurons:
        # -Basic units of computation that take inputs, apply weights and biases, and produce output using an activation function

    # Weights and biases:
        #-weights determine the importance of each input
        # -Bias shifts the output of the activation function
    # Activation Functions:
        # Add non-linearity to the model(eg: ReLU, Sigmoid, Tanh)

# How neural network works:
    # Forward propagation:
        # Data flows through the network to generate predictions
    # Loss Calculation:
        # Compares predictions with actual labels to compute the error
    # BackPropagation:
        # Adjusts weights and biases using Gradient descent to minimizes the loss
```

```python
# Applications of Deep Learning

# Domains and Use Cases:
    # Compute Vision:
        # Image classification (eg: recognizing objects in images)
        # Object detection (eg: autonomous vehicles)
    # Natural Language Processing(NLP):
        # Text Generation(eg GPT model)
        # Sentiment analysis
    # Healthcare:
        # Disease diagnosis (eg identifying tumors in X-rays)
        # Drug discovery
    # Speech Processing:
        # Speech-to-text systems( eg virtual assistants)
        # Voice recognitions
```

```python
from tensorflow.keras.datasets import mnist, cifar10
import tensorflow as tf
import torch 
import torch.nn as nn 
import matplotlib.pyplot as plt


# load MNIST : is Handwritten Datasets 28 x 28 grayscales images
(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = mnist.load_data()
print(f"MNIST Dataset: Train - {X_train_mnist.shape}, Test-{X_test_mnist.shape}")

# load CIFAR-10: images of 32 x 32 color images
(X_train_cifar, y_train_cifar), (X_test_cifar, y_test_cifar) = cifar10.load_data()
print(f"CIFAR-10 Dataset: Train - {X_train_cifar.shape}, Test-{X_test_cifar.shape}")

# Define a basic dense layer
layer = tf.keras.layers.Dense(units=10, activation='relu')
print(f"Tensorflow Layer: {layer}")

# define a basic dense layer
layer = nn.Linear(in_features=10, out_features=5)
print(f"Pytorch Layer: {layer}")

# visualize MNIST sample
plt.imshow(X_train_mnist[0], cmap='gray')
plt.title(f"MNIST label: {y_train_mnist[0]}")
plt.show()

# visualize cifar-10 sample
plt.imshow(X_train_cifar[0])
plt.title(f"CIFAR-10 label: {y_train_cifar[0]}")
plt.show()

```
