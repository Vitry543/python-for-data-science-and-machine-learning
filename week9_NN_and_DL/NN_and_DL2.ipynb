```python
# Forward propagation:
    # -process by which input data flows through the layers of a neural network to produce an output

    # Input layer:
        # -Accepts input features and passes them to the next layer.
    # Hidden layer:
        # -Computes weighted sum of inputs, apply biases, and pass the results through activation function.
    # Output layer:
        # -Produces predictions, typically using an activation function suitable for the task (e.g., softmax for classification).

# Steps in Forward Propagation:
    # 1.Compute weighted sum:
        # z = W.X + b
        # where W is the weight matrix, X is the input vector, and b is the bias vector.
    # 2.Apply activation function:
        # a = activation_function(z)
        # where a is the output of the activation function.
    # 3.Repeat for each layer:
        # output of one layer becomes input for the next layer until the output layer is reached.

# Common activation Functions:
    # 1.Sigmoid:
        # -Maps input to a value between 0 and 1.
        # -Used for binary classification problems.
        # Limitation: Can suffer from vannishing gradients for large positive or negative z.
    
    # 2.Tanh (Hyperbolic Tangent):
        # -Maps input to a value between -1 and 1.
        # -Used for hidden layers in neural networks. or used for hidden layers where zero-centered outputs are preferred.
        # Limitation: Also prone to vanishing gradients for large positive or negative z.

    # 3.ReLU (Rectified Linear Unit):
        # -Outputs the input directly if it is positive; otherwise, it outputs zero.
        # -Widely used in hidden layers of deep neural networks. or used in hidden layers due to simplicity and efficiency
        # Limitation: Can suffer from "dying ReLU" problem where neurons can become inactive and only output zero.

    # 4.Softmax:
        # -Converts a vector of values into a probability distribution.
        # -Used in the output layer for multi-class classification problems.
```

```python
# Choosing Activation functions:

#  Layer type:                             Recommended Activation functions

#  1.Hidden Layers                          ReLU, Tanh, or Leaky ReLU
#  2.Output Layer (Binary Classification)          Sigmoid
#  3.Output Layer (Multi-class Classification)     Softmax
#  4.Output Layer (Regression)                    None or Linear (no activation function) 
```

```python
import numpy as np

# define activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

def softmax(z):
    exp_z = np.exp(z - np.max(z))
    return exp_z/np.sum(exp_z,axis=0,keepdims=True)

# Forward pass function
def forward_pass(X, weights, biases, activation_function):
    z = np.dot(weights, X) + biases
    a = activation_function(z)
    return a

# example imputs
X = np.array([[0.5], [0.2]])  # input features
weights = np.array([[0.1, 0.2],
                    [0.4, 0.5]])  # weights
biases = np.array([[0.1], [0.2]])  # biases

# perform forward pass with different activation functions
activations = {
    'Sigmoid': sigmoid,
    'Tanh': tanh,
    'ReLU': relu,
    'Softmax': softmax
}

for name, func in activations.items():
    output = forward_pass(X, weights, biases, func)
    print(f"{name} activation output:\n{output}\n")
```

```python
import matplotlib.pyplot as plt

# define range of inputs
z = np.linspace(-10, 10, 100)

# plot activation functions
plt.figure(figsize=(12, 8))
plt.plot(z, sigmoid(z), label='Sigmoid', color='blue')
plt.plot(z, tanh(z), label='Tanh', color='green')
plt.plot(z, relu(z), label='ReLU', color='red')
plt.plot(z, softmax(z), label='Softmax', color='orange')
plt.title('Activation Functions')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```
