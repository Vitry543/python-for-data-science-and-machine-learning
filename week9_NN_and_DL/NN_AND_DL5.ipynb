```python
# Introduction to TensorFlow and Keras

# TensorFlow:
    # -Open-source library for numerical computation and ML.
    # -Provides tools for building and training deep learning models.

# Keras:
    # -High-level API integrated with TensorFlow that simplifies the process of creating and training neural network

    # key Features of Keras:
        # -User-Friendly: Intuitive syntax for rapid prototyping.
        # -Modular: Building blocks for defining layers, optimizers, and loss function
        # -Integration: Compatible with Tensorflow for scalable deep learning tasks.
```

```python
# Defining Layers, Models, and Compiling Networks in Keras

# Layers:
    # -Layers are the building block of neural networks. They process input data and pass it through the network.
    # Common Types include:
        # 1.Dense(Fully connected) layers: Each neuron is connected to every neuron in the previous layer.
        # 2.Dropout layers: Randomly drops connections to prevent overfitting.
        # 3.Activation layer: Apply activation functions to introduce non-linearity.

# Models:
    # -Models are a collection of layers that define the architecture of a neural network.
    # 
    # -Keras support two primary ways to define models
        # 1.Sequential API: A linear stack of layers.
        # 2.Functional API: More flexible way to define complex models with multiple inputs and outputs.
# Compiling a Model:
    # specifies:
     # 1. Optimizer: Algorithm used to update weights during training (e.g., Adam, SGD).
     # 2. Loss function: Measures how well the model performs (e.g., categorical_crossentropy for multi-class classification).
     # 3. Metrics: Additional performance measures (e.g., accuracy).
```

```python
# Training, evaluating and saving a Model

# Training:
    # Fit the model to data using model.fit() method, which takes training data, labels, batch size, and number of epochs as arguments.

# Evaluation:
    # Test the model on unseen data using model.evaluate() method, which returns the loss and metrics specified during compilation.

# Saving and loading:
    # Save a trained model using model.save() and reload it with keras.models.load_model() for future use or deployment.
```

```python
# Exercise:
# Objective:
    # Build train, evaluate and save a simple neural network to classify digits from the MNIST dataset
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout

# Dense is a fully connected layer where each neuron is connected to every neuron in the previous layer.
    # -Dense(units, activation) where units is the number of neurons in the layer and activation is the activation function to use (e.g., 'relu' for Rectified Linear Unit).

# Flatten is a multi-dimensional input into single vector for fully connected layers.
# Flatten is used to convert the 2D image data into a 1D vector before feeding it into the Dense layer.

# Conv2D means 2D Convolutional layer that applies filters to input image to extract features. 
    #-Conv2D(num_filters, kernel_size, activation) where num_filters is the number of filters to apply, kernel_size is the size of the filter, and activation is the activation function to use.
    # -Ex:Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)), where 32 filters of size 3x3 are applied to the input image of shape 28x28 with 1 channel (grayscale).


# MaxPooling 2D reduces the spatial dimensions of the input while retaining the most important features by taking the maximum value in each pooling window.
    # -MaxPooling2D(pool_size) where pool_size is the size of the pooling window (e.g., (2, 2) for a 2x2 window).

# Dropout is a regularization technique to randomly set a fraction of input units to zero during training to prevent overfitting.
    # -Dropout(rate) where rate is the fraction of input units to drop (e.g., 0.25 for 25% dropout).

# load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize data
X_train = X_train.reshape(-1, 28, 28, 1).astype('float32')/255.0
X_test = X_test.reshape(-1, 28, 28, 1).astype('float32')/255.0

# one hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

print(f"Training Data Shape: {X_train.shape}")
print(f"Testing Data Shape: {X_test.shape}")

# Build the model
model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)), #Conv2D(num_filters, kernel_size, activation, input_shape) where num_filters is the number of filters to apply, kernel_size is the size of the filter, activation is the activation function to use and input_shape is the shape of the input data.
    MaxPooling2D(pool_size=(2, 2)), #MaxPooling2D(pool_size) where pool_size is the size of the pooling window (e.g., (2, 2) for a 2x2 window).
    Dropout(0.25), # randomly drop 25% of the neurons to prevent overfitting
    Flatten(), # flatten the 2D output from the convolutional layers into a 1D vector
    Dense(128, activation='relu'), # fully connected layer with 128 neurons and ReLU activation
    Dropout(0.5), # randomly drop 50% of the neurons to prevent overfitting
    Dense(10, activation='softmax') #again fully connected layer with 10 neurons (one for each class) and softmax activation for multi-class classification
])

# Display model architecture
print(f"model arcitecture summary: {model.summary()}")

# compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# train the model
histroy = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")

# Save the model and reload
model.save('mnist_classifier.h5')

# load the model
from tensorflow.keras.models import load_model
loaded_model = load_model('mnist_classifier.h5')

# verify loaded model performance
loss, accuracy = loaded_model.evaluate(X_test, y_test)
print(f"Loaded Model Test Loss: {loss}")
print(f"Loaded Model Test Accuracy: {accuracy}")
```
