```python
# integrals : integration is the reverse of differentiation
# integration is the area under the curve, representation of the antiderivative, entire accumation of area under the curve

# the definite integral is the area under the curve between two points is the antiderivative evaluated at the upper limit minus the antiderivative evaluated at the lower limit

# definite integral of f(x) from a to b: int_a^b f(x) dx

# applications of integrals: 
# 1. finding the area under a curve
# 2. finding the volume of a solid of revolution
# 3. finding the length of a curve
# 4. finding the surface area of a solid of revolution
# 5. finding the work done by a force

# application in ML:
# 1.probability distribution
# 2.cost function
# 3.


# integration of f(x) dx = F(x) + C
# integration of f(x2) dx = F(x2) + C
# integration of f(x**3) dx = F(x**4) / 4 + C

# trignometric integrals
# sinx dx = -cosx + C
# cosx dx = sinx + C
# sec^2 x dx = tanx + C
# csc^2 x dx = -cotx + C
# secx tanx dx = secx + C
# cscx cotx dx = -cscx + C

# integration by parts
# integration of u dv = uv - integration of v du
# integration of ln(x) dx = x ln(x) - x + C
# integration of e^x dx = e^x + C
# integration of 1/x dx = ln(x) + C

# partial integration
#partial integration is the integration of a product of functions

# inverse trignometric integrals
# sin^(-1) x dx = x sin^(-1) x + sqrt(1 - x^2) + C
# cos^(-1) x dx = x cos^(-1) x - sqrt(1 - x^2) + C
# tan^(-1) x dx = x tan^(-1) x - (1/2) ln(1 + x^2) + C
# cot^(-1) x dx = x cot^(-1) x + (1/2) ln(1 + x^2) + C
# sec^(-1) x dx = x sec^(-1) x - ln(x + sqrt(x^2 - 1)) + C
# csc^(-1) x dx = x csc^(-1) x + ln(x + sqrt(x^2 - 1)) + C

# hyperbolic integrals
# sinh x dx = cosh x + C
# cosh x dx = sinh x + C
# sech^2 x dx = tanh x + C
# csch^2 x dx = -coth x + C
# sech x tanh x dx = sech x + C
# csch x coth x dx = -csch x + C

```

```python

# optimization concepts
# 1.Local vs Global Minima
# -local minima is the minimum value of a function in a neighborhood of a point
# -global minima is the minimum value of a function in the entire domain of the function

# 2.Convex function: convex functions have a single global minimum
# f(lamba*1+(1-lamba)*2)<=lambda*f(1)+(1-lamba)*f(2) for all lambda in [0,1]

# 3. non-convex function: non-convex functions can have multiple local minima
# f(lamba*1+(1-lamba)*2)>lambda*f(1)+(1-lamba)*f(2) for all lambda in [0,1]
# -most of the machine learning problems are non-convex
# -most neural networks are non-convex and loss functions are non-convex
```

```python
# stochastic gradient descent(SGD) and its Variants
# Stochastic gradient descent(SGD) is a variant of gradient descent that updates the model parameters using a single training example at a time.
# SGD is also an optimization algorithm that is used to minimize the loss function in machine learning models.

# SGD is a optimization algorithm that uses random subset of the data to compute the gradient and update the model parameters.

# why use SGD?
# 1. SGD is a optimization algorithm that uses random subset of the data to compute the gradient and update the model parameters.
# 2. faster convergence for large datasets. compared to batch gradient descent, SGD converges faster for large datasets.

# Variants of SGD
# 1. Mini-batch gradient descent:mini-batch gradient descent is a variant of SGD that uses a small subset of the data to compute the gradient and update the model parameters.
# 2. momentum:momentum is a variant of SGD that uses a moving average of the gradient to update the model parameters.
# 3. Adam Optimizer: it is a variant of SGD that uses a moving average of the gradient and the second moment of the gradient to update the model parameters.

```

```python
# implement SGD for for a linear Model
import numpy as np

# generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# add bias term to X
X_b = np.c_[np.ones((100, 1)), X]

#SGD implementation
def StochasticGradientDescent(X, y, theta, learning_rate, num_iterations):
    m = len(y)
    for i in range(num_iterations):
        for j in range(m):
            random_index = np.random.randint(m)
            xi = X[random_index:random_index+1]
            yi = y[random_index:random_index+1]
            # gradient = 2 * xi.T.dot(xi.dot(theta) - yi)
            gradient = 2 * xi.T @ (xi.dot(theta) - yi)
            
            theta = theta - learning_rate * gradient
    return theta 

# initialize parameters
theta = np.random.randn(2, 1)

# define learning rate
learning_rate = 0.1

# define number of iterations
num_iterations = 1000

# run SGD
theta = StochasticGradientDescent(X_b, y, theta, learning_rate, num_iterations)

# print final parameters
print("Final parameters:", theta)

import matplotlib.pyplot as plt
plt.scatter(X, y, color="blue", label="Linear Regression")
plt.plot(X, X_b.dot(theta), "r-")
plt.title("Linear Regression")
plt.show()

```
