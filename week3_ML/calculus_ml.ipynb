```python
# partial Derivatives
# partial derivatives are used to find the rate of change of a function with respect to one of its inputs while holding the other inputs constant

# for a function f(x, y) = x^2 + y^2, the partial derivative with respect to x is f_x(x, y) = 2x, and the partial derivative with respect to y is f_y(x, y) = 2y

# gradient: vector of all partial derivatives of a function, indicating the direction of steepest ascent

# for a function f(x, y) = x^2 + y^2, the gradient is grad f(x, y) = (2x, 2y)

# chain rule: used to find the derivative of a composite function

# for a function f(g(x)) = x^2 + y^2, the derivative is f'(g(x)) = 2x + 2y
```

```python
x,y=sp.symbols('x y')
f=x**2+y**2
partial_x=sp.diff(f,x)
partial_y=sp.diff(f,y)
print(partial_x)
print(partial_y)

# gradient
gradient=sp.Matrix([partial_x,partial_y])
print(gradient)
```

```python
# implement gradient descent for linear regression
import numpy as np
import matplotlib.pyplot as plt

# define the gradient descent function
def gradient_descent(x,y, theta, learning_rate, iterations):
    m=len(y)
    for _ in range(iterations):
        predictions =np.dot(x,theta)
        errors=predictions-y
        gradient=(1/m)*np.dot(x.T,errors)
        theta=theta-learning_rate*gradient
    return theta

# sample data
x=np.array([[1,1],[1,2],[1,3]])
y=np.array([2,2.5,3.5])

# initialize parameters
theta=np.array([0.1,0.1])
learning_rate=0.1
iterations=1000

# run gradient descent
optimized_theta=gradient_descent(x,y,theta,learning_rate,iterations)

print("Optimized theta:",optimized_theta)

plt.scatter(x[:,1],y)
predictions=np.dot(x,optimized_theta)
plt.plot(x[:,1],predictions)
plt.show()

    
```

```python
# implement gradient descent with multiple learning rates and compare convergence speeds

import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, theta,learning_rate, num_iterations):
    m = len(y)
    cost_history = np.zeros(num_iterations)
    for i in range(num_iterations):
        predictions = np.dot(x, theta)
        errors = predictions - y
        gradient = np.dot(x.T, errors) / m
        theta = theta - learning_rate * gradient
        cost_history[i] = np.dot(errors.T, errors) / (2 * m)
    return theta,cost_history

# generate some data
x = np.array([[1,1],[1,2],[1,3],[1,4],[1,5]])
y = np.array([2,2.5,3,3.5,4])


# initialize parameters
theta = np.array([0.1, 0.1])
learning_rate = 0.1
num_iterations = 1000

# run gradient descent
optimized_theta = gradient_descent(x, y, theta, learning_rate, num_iterations)
print("Optimized theta:", optimized_theta)

# run gradient descent with different learning rates
learning_rates = [0.1, 0.01, 0.001]
for learning_rate in learning_rates:
    optimized_theta, cost_history = gradient_descent(x, y, theta, learning_rate, num_iterations)
    plt.plot(cost_history,label=f'Learning Rate = {learning_rates}', color='skyblue', marker='o', linestyle='-')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Gradient Descent with Different Learning Rates')
plt.show()

# convergence speed
def convergence_speed(cost_history):
    return cost_history[-1] - cost_history[0]

convergence_speed(cost_history)
plt.plot(cost_history, label='Convergence Speed', color='skyblue', marker='o', linestyle='-')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Convergence Speed')
plt.show()
```

```python
# implement gradient descent with multiple learning rates and compare convergence speeds

import numpy as np
import matplotlib.pyplot as plt

# define the function to optimize
f = lambda x: x**2 + 3*x + 2

# define the gradient of the function
def gradient(x):
    return 2*x + 3

# define the learning rates
learning_rates = [0.1, 0.01, 0.001]

# define the number of iterations
num_iterations = 100

# initialize the starting point
x = 0

# run gradient descent for each learning rate
for learning_rate in learning_rates:
    # store the values of x and f(x) for each iteration
    x_values = []
    f_values = []
    
    # run gradient descent
    for i in range(num_iterations):
        x_values.append(x)
        f_values.append(f(x))
        x = x - learning_rate * gradient(x)
    
    # plot the results
    plt.plot(x_values, f_values, label=f"learning rate = {learning_rate}")

# add labels and legend
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# visualize the gradient descent process on a quadratic function

import numpy as np
import matplotlib.pyplot as plt

A=np.array([[1,1],[1,2],[1,3],[1,4],[1,5]])
b=np.array([2,2.5,3,3.5,4])

def gradient_descent(A,b,theta,learning_rate,num_iterations):
    m=len(b)
    cost_history=np.zeros(num_iterations)
    for i in range(num_iterations):
        predictions=np.dot(A,theta)
        errors=predictions-b
        gradient=np.dot(A.T,errors)/m
        theta=theta-learning_rate*gradient
        cost_history[i]=np.dot(errors.T,errors)/(2*m)
    return theta,cost_history

# initialize parameters
theta=np.array([0.1,0.1])
learning_rate=0.1
num_iterations=1000

# run gradient descent
optimized_theta,cost_history=gradient_descent(A,b,theta,learning_rate,num_iterations)
print("Optimized theta:",optimized_theta)

# plot the cost history
plt.plot(cost_history)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Gradient Descent')
plt.show()

# for quadratic function
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()
```
