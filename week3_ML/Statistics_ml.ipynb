```python
#Measure of central tendency and Dispersion

# central tendency
# 1. mean
# 2. median
# 3. mode

# Dispersion
# 1. range
# 2. variance
# 3. standard deviation
# 4. interquartile range
# 5. skewness
# 6. kurtosis
```

```python
# Central tendency
import statistics
from statistics import mode

# mean is the average of the data
data = [10,20,30,40,50,50,50]
mean = sum(data)/len(data)
print("mean",mean)

# median is the middle value in the data
median = sorted(data)[len(data)//2] if len(data) % 2 == 1 else \
     (sorted(data)[len(data)//2] + sorted(data)[len(data)//2 - 1]) / 2
print("median",median)

# mode is the most common value in the data
print("mode: ",mode(data))
mode = mode(data)
print("mode",mode)

# Dispersion
# variance is the average of the squared differences from the mean

variance = sum((x - mean) ** 2 for x in data) / len(data)
print("variance",variance)


# standard deviation is the square root of the variance
std_dev = variance ** 0.5
print("std_dev",std_dev)

import math
std_dev = math.sqrt(variance)
print("std_dev",std_dev)

```

```python
# confidence intervals and statistical Significance
# confidence Interval : range of Values within which the true population parameter is excepted to lie:
# CI=X bar +/- z*(s/n**0.5) , X is sample means, z= Z-score , S- Standard deviation

```

```python
import scipy.stats as stats
import numpy as np
data = [10,20,30,40,50,50,50]
mean = sum(data)/len(data)
variance = sum((x - mean) ** 2 for x in data) / len(data)
std_dev=variance**0.5

sample_mean=mean
z_score=1.96

ci=(sample_mean-z_score * (std_dev/np.sqrt(len(data))),
    sample_mean + z_score * (std_dev/np.sqrt(len(data))))

print("95% confidence interval:",ci)
```

```python
# perform a T-test to compare the means of two groups
from scipy.stats import ttest_ind
import numpy as np


# generate two random samples
group1 = np.random.normal(10, 2, 100)
group2 = np.random.normal(12, 2, 100)

# perform a T-test
t_stat, p_Value = ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("p-value:", p_Value)

# interpretation
alpha=0.05
if p_Value< alpha:
    print("reject null hypothesis: means are significantly different")
else:
    print("fail to reject null hypothesis: means are not significantly different")
```
