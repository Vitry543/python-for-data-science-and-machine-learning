```python
# applying linear algebra, calcilus, and statistics to real world problems

# linear algebra is used to represent and manipulate data
# calcilus is used to find patterns and relationships in data
# statistics is used to make predictions and decisions based on data

# Linear algebra:
# -Mathematical language for data
# -Represents data as vectors and matrices
# -Operations on vectors and matrices

# - Mathematical Model: y_hat =x*theta, where y_hat is the predicted value, x is the feature matrix(with bias term), and theta is the parameter(weight and bias)


# Calculus:
# - Optimization: Minimizing the cost function
# -optimization of theta involves minimizing the cost function
# J(theta) = 1/2m * summation((y_hat(i) - y(i))^2) for i=1 to m

# gradient of J(theta) = 1/m * X.T * (X*theta - y) 

# Statistics:
# - Metrics like mean Squared Error(MSE) and R-squared are used to evaluate the performance of the model
```

```python
# 1.using Gradient Descent for parameter optimization
# Gradient Descent is an optimization algorithm that is used to find the minimum of a function.
# Algorithm:
# 1. Initialize the parameters
# 2. Compute the gradient of the function
# 3. Update the parameters
# 4. Repeat steps 2 and 3 until the parameters converge

# under Gradient Descent Algorithm
# -iterativel upadeta Theta using gradient of the function formula theta=theta-alpha*gradient of J
# where alpha is the learning rate


# 2.evaluate the model using statistics metrics
# - Mean Square Error (MSE): MSE=1/m * summation(y_hat(i)-y(i))^2 for i =1 to m
# where y_hat is the predicted value and y is the actual value, measures the average of the squares of the errors
# - R-Squared: R^2=1-(ssr/sst), where ssr: sum of squares of residuals, sst: sum of squares of total
#- where it measures how well the regression line explains the variation in the data 

```

```python
# Task1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def predict(x, theta):
    return np.dot(x, theta)

# task2

def gradient_descent(x, y, theta, learning_rate, num_iterations):
    m = len(y)
    for i in range(num_iterations):
        y_pred = predict(x, theta)
        error = y_pred - y
        gradient = np.dot(x.T, error) / m
        theta = theta - learning_rate * gradient
    return theta

# task3: find the mean squared error

def mean_squared_error(y, y_pred):
    return np.mean((y - y_pred) ** 2)

def r_squared(y, y_pred):
    ssr = np.sum((y - y_pred) ** 2)
    sst = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ssr / sst)

x = np.array([[1, 1], [1, 2], [1, 3], [1, 4], [1, 5]])
y = np.array([2, 4, 6, 8, 10])
theta = np.zeros(2)
learning_rate = 0.01
num_iterations = 1000
theta = gradient_descent(x, y, theta, learning_rate, num_iterations)
predicted_y = predict(x, theta)

print(predicted_y)
print(theta)
print(mean_squared_error(y, predicted_y))
print(r_squared(y, predicted_y))
```
