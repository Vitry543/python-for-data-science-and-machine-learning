````python
# Creating and visualizing a multinomial distribution is a great way to understand probabilities for multi-class data, especially when you have a fixed number of trials ($n$) and the outcomes of each trial are independent, belonging to one of $K$ categories with fixed probabilities ($\mathbf{p} = [p_1, p_2, \dots, p_K]$).

# While the **Normal**, **Poisson**, and **Binomial** distributions are fundamental, they primarily deal with:
# * **Normal:** Continuous data, often modeling sums or averages.
# * **Poisson:** Counts of events in a fixed interval (one dimension).
# * **Binomial:** Success/Failure outcomes (two dimensions: successes and failures).

# The **Multinomial Distribution** is the generalization of the Binomial distribution to more than two outcomes ($K > 2$).

# Here is a complete Python example using `numpy` for calculation and `matplotlib` for visualization.

### 1. The Concept: Multinomial Distribution

# The probability mass function (PMF) for the multinomial distribution is:

# $$P(X_1=k_1, X_2=k_2, \dots, X_K=k_K) = \frac{n!}{k_1! k_2! \dots k_K!} p_1^{k_1} p_2^{k_2} \dots p_K^{k_K}$$

# where:
# * $n$ is the total number of trials.
# * $K$ is the number of possible outcomes (classes).
# * $k_i$ is the observed count for outcome $i$, such that $\sum_{i=1}^K k_i = n$.
# * $p_i$ is the probability of observing outcome $i$, such that $\sum_{i=1}^K p_i = 1$.



### 2. Python Implementation and Visualization

# For this example, let's assume we have **3 classes ($K=3$)** and we perform **$n=20$ trials**.


import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from scipy.special import factorial # Using scipy for factorials as they grow very fast

# --- 1. Define Parameters ---
N_TRIALS = 20  # Total number of trials (n)
K_CLASSES = 3  # Number of categories (K)

# Fixed probabilities for each class (must sum to 1)
P_TRUE = np.array([0.2, 0.5, 0.3]) 

print(f"Parameters: N={N_TRIALS}, K={K_CLASSES}")
print(f"True Probabilities (p): {P_TRUE}")
print("-" * 30)

# --- 2. Simulate Multinomial Data ---
# We use numpy's built-in multinomial function to generate random samples.
N_SAMPLES = 10000  # Number of times we repeat the experiment (N trials)
simulated_data = np.random.multinomial(N_TRIALS, P_TRUE, size=N_SAMPLES)

# simulated_data is an array where each row is one experiment result: [k1, k2, k3]

# --- 3. Analyze and Visualize the Simulation Results ---

# A) Calculate the observed mean counts across all samples
observed_mean_counts = np.mean(simulated_data, axis=0)

# B) Calculate the expected mean counts (The mean of a multinomial distribution is n * p)
expected_mean_counts = N_TRIALS * P_TRUE

print(f"Expected Mean Counts (n*p): {expected_mean_counts}")
print(f"Observed Mean Counts (from {N_SAMPLES} simulations): {observed_mean_counts.round(2)}")
print("-" * 30)

# --- 4. Visualization: Comparing Expected vs. Observed Frequencies ---

class_labels = [f'Class {i+1}' for i in range(K_CLASSES)]
x_pos = np.arange(K_CLASSES)

plt.figure(figsize=(10, 6))

bar_width = 0.35

# Bar plot for Observed Frequencies
plt.bar(x_pos - bar_width/2, 
        observed_mean_counts, 
        bar_width, 
        label='Observed Mean Counts', 
        color='skyblue')

# Bar plot for Expected Frequencies
plt.bar(x_pos + bar_width/2, 
        expected_mean_counts, 
        bar_width, 
        label='Expected Mean Counts (n*p)', 
        color='salmon',
        alpha=0.7)

plt.xlabel('Outcome Class')
plt.ylabel(f'Average Count (out of {N_TRIALS} trials)')
plt.title(f'Multinomial Distribution Analysis (N={N_TRIALS}, Samples={N_SAMPLES})')
plt.xticks(x_pos, class_labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# --- 5. Visualization: Exploring the Distribution of One Class ---

# Let's see how the count for Class 2 (which has p=0.5) is distributed across the samples
class_2_counts = simulated_data[:, 1] # Get the second column

plt.figure(figsize=(8, 5))
plt.hist(class_2_counts, bins=np.arange(0, N_TRIALS + 2) - 0.5, density=True, 
         edgecolor='black', alpha=0.7, color='green')

# Overlay the theoretical Binomial distribution for this specific class
# Note: If you fix n and p_i, the count k_i follows a Binomial(n, p_i) distribution.
from scipy.stats import binom
x_values = np.arange(0, N_TRIALS + 1)
pmf_binomial = binom.pmf(x_values, N_TRIALS, P_TRUE[1])

plt.plot(x_values, pmf_binomial, 'r-o', label=f'Theoretical Binomial P(Class 2)')

plt.axvline(expected_mean_counts[1], color='k', linestyle='--', 
            label=f'Expected Mean ({expected_mean_counts[1]:.2f})')

plt.title(f'Distribution of Counts for Class 2 (p={P_TRUE[1]})')
plt.xlabel('Number of times Class 2 occurred (k_i)')
plt.ylabel('Probability Density / Frequency')
plt.xticks(np.arange(0, N_TRIALS + 1, 2))
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


# ### Explanation of the Output and Connection to Other Distributions

# 1. **Simulation Setup:** We defined $n=20$ trials and three classes with probabilities $P=[0.2, 0.5, 0.3]$. We ran this experiment 10,000 times (`N_SAMPLES`).

# 2. **Bar Chart Visualization (Multinomial):**
#    * This chart compares the **Expected Mean Counts** (which are simply $n \times p_i$) with the **Observed Mean Counts** from the simulation.
#    * As predicted by the Law of Large Numbers, the observed means closely track the expected means, confirming the simulation of the multinomial process.

# 3. **Histogram Visualization (Connection to Binomial):**
#    * This is the crucial link: **The marginal distribution of any single outcome count ($k_i$) from a Multinomial distribution is a Binomial distribution.**
#    * For example, if we only cared about whether Class 2 occurred (a success) or didn't occur (a failure), that single dimension would follow a **Binomial distribution** with parameters $n=20$ and $p=0.5$.
#    * The histogram of the counts for Class 2 perfectly overlays the theoretical Binomial PMF (Probability Mass Function), demonstrating this relationship.

# **In summary:**

# * **Multinomial:** Models the outcome of $n$ trials across $K$ categories simultaneously.
# * **Binomial:** Is the special case of Multinomial where $K=2$ (Success/Failure), or the marginal distribution of any single category within a larger Multinomial setup.
# * **Normal/Poisson:** Are not directly used here as the multinomial is discrete, but the Normal distribution can approximate the Multinomial distribution when $n$ is very large (via the Central Limit Theorem).
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Parameters
lambda_per_hour = 120  # Average calls per hour
hours = 24

# Simulate calls per hour
np.random.seed(42)
calls_per_hour = np.random.poisson(lambda_per_hour, hours)

# Create time series DataFrame
df_calls = pd.DataFrame({
    'hour': np.arange(1, hours + 1),
    'calls': calls_per_hour
})

# Analysis & Visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Time series of call volume
axes[0, 0].plot(df_calls['hour'], df_calls['calls'], 
                marker='o', linewidth=2, markersize=6, color='steelblue')
axes[0, 0].axhline(y=lambda_per_hour, color='red', linestyle='--', 
                   label=f'Theoretical Mean ({lambda_per_hour})')
axes[0, 0].set_title('Call Volume Over 24 Hours', fontsize=13, fontweight='bold')
axes[0, 0].set_xlabel('Hour')
axes[0, 0].set_ylabel('Number of Calls')
axes[0, 0].legend()
axes[0, 0].grid(alpha=0.3)

# Plot 2: Histogram vs Theoretical PMF
observed_counts = np.bincount(calls_per_hour)
unique_values = np.arange(len(observed_counts))
axes[0, 1].bar(unique_values, observed_counts/hours, 
               alpha=0.7, color='steelblue', edgecolor='black', label='Observed')
axes[0, 1].plot(unique_values, stats.poisson.pmf(unique_values, lambda_per_hour), 
                'ro-', label='Poisson PMF', linewidth=2)
axes[0, 1].set_title('Observed vs Theoretical Distribution', fontsize=13, fontweight='bold')
axes[0, 1].set_xlabel('Calls per Hour')
axes[0, 1].set_ylabel('Probability')
axes[0, 1].legend()
axes[0, 1].grid(alpha=0.3)

# Plot 3: Descriptive statistics
stats_text = f"""
Mean: {calls_per_hour.mean():.2f}
Std Dev: {calls_per_hour.std():.2f}
Variance: {calls_per_hour.var():.2f}
Min: {calls_per_hour.min()}
Max: {calls_per_hour.max()}
"""
axes[1, 0].text(0.1, 0.5, stats_text, fontsize=12, 
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue"))
axes[1, 0].set_title('Summary Statistics', fontsize=13, fontweight='bold')
axes[1, 0].axis('off')

# Plot 4: QQ Plot
stats.probplot(calls_per_hour, dist=stats.poisson(lambda_per_hour), plot=axes[1, 1])
axes[1, 1].set_title('QQ Plot: Goodness-of-Fit', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()

# Statistical test
chi2_stat, p_value = stats.kstest(calls_per_hour, 'poisson', args=(lambda_per_hour,))
print(f"KS Test p-value: {p_value:.4f}")
print(f"Data {'matches' if p_value > 0.05 else 'does not match'} Poisson distribution")
````

````python
# calculate the probability of the flip a coin using bayes theorem

def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    """
    Calculates P(A|B) using Bayes' Theorem.
    
    Formula: P(A|B) = (P(B|A) * P(A)) / P(B)
    
    Args:
        p_a: Prior probability of hypothesis A being true. P(A)
        p_b_given_a: Probability of evidence B given A is true. P(B|A)
        p_b_given_not_a: Probability of evidence B given A is false. P(B|~A)
    
    Returns:
        Posterior probability P(A|B)
    """
    # Calculate marginal probability of Evidence B: P(B)
    # P(B) = P(B|A)*P(A) + P(B|~A)*P(~A)
    p_not_a = 1 - p_a
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    
    # Calculate posterior P(A|B)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    
    return p_a_given_b

def main():
    # --- Configuration ---
    # Hypothesis A: The coin is Biased.
    # Hypothesis ~A: The coin is Fair.
    
    # Properties of the coins:
    prob_heads_biased = 0.9  # Biased coin lands Heads 90% of the time
    prob_heads_fair = 0.5    # Fair coin lands Heads 50% of the time
    
    # Initial Belief (Prior):
    # We pick a coin from a bag where 50% are biased and 50% are fair.
    p_biased_prior = 0.5
    
    print(f"Initial belief (Prior) that coin is biased: {p_biased_prior:.4f}")
    print("-" * 50)

    # --- Simulation ---
    # Let's observe a sequence of flips: Heads, Heads, Tails
    flips = ['Heads', 'Heads', 'Tails']
    
    current_belief = p_biased_prior

    for i, result in enumerate(flips, 1):
        # Determine likelihoods based on the flip result
        if result == 'Heads':
            likelihood_biased = prob_heads_biased
            likelihood_fair = prob_heads_fair
        else: # Tails
            likelihood_biased = 1 - prob_heads_biased
            likelihood_fair = 1 - prob_heads_fair
            
        # Update belief using Bayes Theorem
        current_belief = bayes_theorem(current_belief, likelihood_biased, likelihood_fair)
        
        print(f"Flip {i}: {result}")
        print(f"  -> Updated belief (Posterior) that coin is biased: {current_belief:.4f}")

    # --- Prediction ---
    # Calculate probability of the NEXT flip being Heads based on current belief
    # P(Next=H) = P(H|Biased)*P(Biased) + P(H|Fair)*P(Fair)
    p_next_heads = (prob_heads_biased * current_belief) + (prob_heads_fair * (1 - current_belief))
    
    print("-" * 50)
    print(f"Based on observed data, probability that the NEXT flip is Heads: {p_next_heads:.4f}")

if __name__ == "__main__":
    main()
````
