```python
# Feature Engineering:
    # -Process of transforming raw data into features that better represent the underlying problem to the predictive models, resulting in improved model accuracy on the validation set.
    # -Feature Engineering is a process of using domain knowledge of the data to create features that make the machine learning algorithms work better.

# importance of feature Engineering:
    # -improve model accuracy
    # -reduce overfitting
    # -reduce training time
    # -reduce model complexity
    # -enables model Interpretability
    # - handles Data chanllenges

# Key Applications
    # -Finance
    # -Healthcare
    # -Retail
    # -Marketing
    # -Sales
    # -Supply Chain

# Types of feature 
    # 1.categorical feature 
        # -represent discrete categories or labels or groups of data
        # -Encoding Techniques:
            # 1.one hot encoding
            # 2.label encoding
            # 3.target encoding
            # 4.frequency encoding
            # 5.binary encoding
        # 1.1 Nominal feature - no order
        # 1.2 Binary feature - two categories
    # 2.Numerical feature 
        # -represent continuous values or measurements or discrete numbers
        # preproceesing Techniques:
            # -Scaling Techniques:
                # 1.min-max scaling, standard scaling, normalization, log transformation, box-cox transformation
            # 
    # 3.ordinal feature 
        # -represent ordered categories or labels or groups of data
        # Encoding Techniques:
            # 1.ordinal encoding


```

```python
# Overview of Feature Engineering techniques

# Scaling : Ensures all features contribute equally to the model
    # Techniques:
        # 1.min-max scaling
        # 2.standard scaling
        # 3. Standardization

# Encoding: converts categorical variables into numeric format
    # Techniques:
        # 1.one hot encoding
        # 2.label encoding
        # 3.target encoding
        # 4.frequency encoding
        # 5.binary encoding

# Feature Selection: Identifies the most relevant features for the model or reduces the number of features to improve model performance
    # Techniques:
        # 1.Statistical Methods
        # 2. Recursive Feature Elimination(RFE)
        # 3. Lasso Regression
        # 4. Ridge Regression
        # 5. Random Forest

# Transformation: Transforms the data to a different scale or distribution
    # Techniques:
        # 1.log transformation
        # 2.box-cox transformation
        # 3. polynomial Features
        
```

```python
import pandas as pd
import numpy as np

# load the titanic dataset
url='https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# describe the dataset
print("dataset info: \n")
print(df.info())

# preview the first 5 rows
print("\n dataset preview: \n ", df.head())


# seperate features
categorical_features=df.select_dtypes(include=['object']).columns
print("\n categorical features: ", categorical_features)
print("\n categorical features: ", categorical_features.tolist())


numerical_features=df.select_dtypes(include=['int64','float64']).columns
print("\n numerical features:", numerical_features)
print("\n numerical features:", numerical_features.tolist())


# feature engineering

# display summary of categorical features
print("\n categorical features summary: \n")
for col in categorical_features:
    print(f'{col}:\n', df[col].value_counts(),"\n")

# display summary of numerical features
print("\n numerical features summary: \n")
for col in numerical_features:
    print(f'{col}:\n', df[col].value_counts(),"\n")


df.describe()
```
