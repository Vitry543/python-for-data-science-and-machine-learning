```python
# Building CNN architecture with Keras\
# Steps to Build a CNN :
    # -Convolutional layers: Extract features from the input images
    # -Pooling layers: Downsample feature maps to reduce dimensions and retain key features
    # -Dense(fully connected) layers: Combine features for final predictions

# Basic CNN architecture:
    # Input Layer --> Convolutional layer --> Activation --> Pooling --> fully connected layer --> Output layer
    # 
    # -repeat the convolutional and pooling layer for deeper network  
```

```python
# Complining, Training, and Evaluating a CNN
# Steps:
    # Compile the model:
        # -Define loss, Optimizer, and metrics
        # Examples of loss function
            # Categorical Cross-entropy: Mutli-class classifiaction
        # Example of optimizer:
            # Adam: Efficient optimization for large network
        # Example of Metrics: Accuracy
    
    # Train the model:
        # -use model.fit() with training data, validation data, epochs, and batch size

    # Evaluate the model
        # -Use model.evaluate() with test data to calculate metrics
```

```python
# Introduction to popular CNN architectures

# 1.LeNet:
    # -one of the earliest CNNs for handwritten digit classification (eg: MNIST)
# 2.AlexNet:
    # -Revolutionized deep learning for image classification in 2012
    # -Introduced ReLU activation and dropout for regularization
# 3.VGG:
    # -Uses deep network with small features (eg 3x3 3x3)
    # -Known for its simplicity and effectiveness
```

```python
# Exercise: Objective:
    # -Build, train, and evaluate a CNN for image classification on the MNIST or CIFAR-10 dataset using keras and Tensorflow
```

```python
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# - X_train and X_test contain corresponding image data, while y_train and y_test contain the corresponding labels for those images

# Normalize the data
X_train = X_train.astype('float32') / 255.0 
X_test = X_test.astype('float32') / 255.0 

#converts image data --> 32 bit floating point, and by dividing with 255.0 
    # -it normalizes the pixel value from 0 to 255 --> 0 to 1 to improve model convergence during the training process


# One-hot encode the labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
# Converts the integral labels 0 to 9 --> 1 hot encoding format
    # -ex: label three will become 0010000 

print(f"training data shape: {X_train.shape}, label shape: {y_train.shape}")
print(f"testing data shape: {X_test.shape}, label shape: {y_test.shape}")

# build the CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)), # 32 filters of size 3x, relu activation and image shape of 32x32 and RGB channels
    MaxPooling2D((2,2)), #adds 2x2 max pooling layer to reduce the spatial dimensions
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(), #flattens 2D feature map --> 1D aray for connected layers
    Dense(128, activation='relu'), # this creates full connected layer with 128 units and relu activation
    Dropout(0.5), #to prevent overfitting drop 50% dropout rate
    Dense(10, activation='softmax') # Output layer with 10 units, one for each class and softmax activation 
])

# Display the model summary
model.summary()

# Compile the model
model.compile(
    optimizer ='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# TRain the model
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)

# Evaluate the test dataset
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_accuracy: .4f}, test loss:{test_loss :.4f}")

# plot
import matplotlib.pyplot as plt

# plot the accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='validation accuracy')
plt.title("model accuracy")
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend()
plt.show()

# Plot the loss
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.title("model loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```
