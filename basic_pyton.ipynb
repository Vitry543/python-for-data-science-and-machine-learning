Number Guessing Game


---



```python
import random

secret_numebr = random.randint(1, 100)
attempts = 3

print("I am thinking of a number between 1 and 100")

while attempts > 0:
  guess = int(input("Take a guess: "))
  if guess == secret_numebr:
    print("Congratulations! You guessed the number!")
    break
  elif guess < secret_numebr:
    print("Too low! Try again.")
  else:
    print("Too High High! Try again.")
  attempts -= 1

if attempts == 0:
  print("sorry, you've run out of attempts. The secret number was:", secret_numebr)
```

### Basic Math Quiz Game

Let's create a simple math quiz game where you'll be asked a few basic arithmetic questions. The game will keep track of your score.

```python
import random

def generate_question():
  """Generates a random math question (addition, subtraction, or multiplication)."""
  num1 = random.randint(1, 10)
  num2 = random.randint(1, 10)
  operator = random.choice(['+', '-', '*'])
  question = f"What is {num1} {operator} {num2}? "

  if operator == '+':
    answer = num1 + num2
  elif operator == '-':
    answer = num1 - num2
  else: # operator == '*'
    answer = num1 * num2

  return question, answer

generate_question()
```

```python
score = 0
num_questions = 3 # You can change the number of questions

print("Welcome to the Basic Math Quiz!")
print(f"You will be asked {num_questions} questions.")

for i in range(num_questions):
  question, correct_answer = generate_question()

  while True:
    try:
      user_answer = int(input(question))
      break
    except ValueError:
      print("Invalid input. Please enter a number.")

  if user_answer == correct_answer:
    print("Correct! 🎉")
    score += 1
  else:
    print(f"Incorrect. The correct answer was {correct_answer}. 😔")

print("\n--- Quiz Over ---")
print(f"You got {score} out of {num_questions} questions correct.")
```

Shopping List App

---



```python
# Step 1: Initialize the empty shopping list
shopping_list = []

# Step 2: Define the main menu
def show_menu():
  print("\n --- Shopping List App ---")
  print("1. View the shopping list")
  print("2.Add an Item")
  print("3.Remove an item")
  print("4.Clear the List")
  print("5. Quit")

# Step 3: Main program Loop
while True:
  show_menu()
  choice = input("Select an option (1/2/3/4/5): ")
  if choice == '1':
    print("\n--- Shopping List ---")
    if not shopping_list:
      print("your shopping list is empty")
    else:
      for index, item in enumerate(shopping_list):
        print(f"{index + 1}. {item}")

  elif choice == '2':
    item = input("enter the item to add: ")
    shopping_list.append(item)
    print(f"{item} has been added to the shopping list")

  elif choice == '3':
    item = input("enter the item to remove: ")
    if item in shopping_list:
      shopping_list.remove(item)
      print(f"{item} has been removed from the shopping list")
    else:
      print(f"{item} is not in the shopping list")

  elif choice == '4':
    shopping_list.clear()
    print("the shopping list has been cleared")

  elif choice == '5':
    print("quitting the program")
    break
  else:
    print("Invalid choice")
```
