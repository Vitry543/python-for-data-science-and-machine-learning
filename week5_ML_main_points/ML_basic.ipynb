```python
# Machine Learning: 
    # -machine learning is a subset of artificial intelligence that involves training models to make predictions or decisions based on data.

# real world applications of machine learning:
    # -image recognition
    # -speech recognition
    # -recommendation systems
    # -healthcare
    # -finance
    # -self driving cars
    # -E-commerce
    # -Natural Language Processing

# ML important : ML is important to understand the data and the problem before applying machine learning.

# Types of ML:
    # -Supervised Learning
    # -Unsupervised Learning
    # -Reinforcement Learning

# Supervised Learning:
    # -In supervised learning, the model is trained on a labeled dataset, where each input is associated with a corresponding output.
    # -The goal is to learn a mapping from inputs(features) to outputs(target), so that the model can make predictions on new, unseen data.
    # -Common examples of supervised learning include regression, classification, and clustering.
# key points:
    # -labeled data is used to train the model
    # -Accuracy depends heavly on the quality of the data


# Unsupervised Learning:
    # -In unsupervised learning, the model is trained on an unlabeled dataset, where the inputs are not associated with any outputs.
    # -Common examples of unsupervised learning include clustering, dimensionality reduction, and anomaly detection.
# key points:
    # -unlabeled data is used to train the model
    # -focused on exploring the data and finding patterns


# Reinforcement Learning:
    # - An agent interacts with an environment and learn by trials and errors to maximize cumulative reward.
    # -In reinforcement learning, the model is trained on a dataset of examples, where each example is associated with a corresponding output.
    # -Common examples of reinforcement learning include classification, regression, and clustering.
    # examples: Robotics| Gaming| dynamic systems

# key points:
    # -goal-oriented learning based on reward-based learning
    # -suitable for sequential decision making problems
```

```python
# key concepts of machine learning:
# 1.Features : the input variables(independent variables) used to train the model
    # examples: age, gender, income, etc.

# 2.Labels : the output variables(dependent variables) used to train the model
    # examples: salary, etc.

# 3.target : the output variable(dependent variable) we want to predict
    # examples: salary,house price etc.

# 4. Training data : the data used to train the model
    # examples: age, gender, income, salary, etc.

# 5. Testing data : the data used to test the model
    # examples: age, gender, income, salary, etc.

# The data is split into training and testing data
# a typical split is 80% training and 20% testing

# 6.overfitting : when the model performs well on the training data but not on the testing data
    # -model learns the noise and details in the training data, performing poorly on new data
    # -model becomes too complex for the dataset

# 7.underfitting : when the model performs poorly on both the training data and the testing data
    # -model is too simple for the dataset
    # -model fails to capture the patterns in the data

# 8.bias-variance tradeoff : the balance between bias and variance
    # -bias : the difference between the expected value and the actual value
        # -the error introduced by assuming a simplified model
    # -variance : the difference between the expected value and the actual value
        # -error introduced by the model's senstivity to small changes in the training data
        
    # goal: balance between bias and variance to achieve good(optimal) performance

```

```python
# 1.define features and target varaibles

url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'

import pandas as pd
 
# load the dataset
df=pd.read_csv(url)

# Define Features and targets
features=df[['total_bill','size']]
target=df['tip']

print("Features: \n",features.head())
print("Target: \n",target.head())


```

```python
# split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train,X_test, Y_train, y_test = train_test_split(features,target,test_size=0.2, random_state=42)

print("Training data Set: ",X_train.shape)
print("Testing data Set: ",X_test.shape)
```

```python
# Visualize the data
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, x_vars=['total_bill', 'size'], y_vars='tip', height=5, aspect=0.8)
plt.title('features vs target realtionships')
plt.legend
plt.show()
```
