```python
# 1.conduct T-test
# perform one-sample, two sample, paired t-tests

from scipy.stats import ttest_1samp, ttest_ind, ttest_rel

# one-sample t-test
sample_data = [1, 2, 3, 4, 5]
population_mean = 3
t_stat, p_value = ttest_1samp(sample_data, population_mean)
print("One-sample t-test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)

# two-sample t-test
group1 = [1, 2, 3, 4, 5]
group2 = [6, 7, 8, 9, 10]
t_stat, p_value = ttest_ind(group1, group2)
print("\nTwo-sample t-test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)

# paired t-test
before_treatment = [1, 2, 3, 4, 5]
after_treatment = [2, 3, 4, 5, 6]
t_stat, p_value = ttest_rel(before_treatment, after_treatment)
print("\nPaired t-test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)


# plotting one-sample t-test results
import matplotlib.pyplot as plt
import numpy as np
x = np.array(sample_data)
plt.hist(x, bins=5, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(np.mean(x), color='red', linestyle='dashed', linewidth=1)
plt.title('One-sample t-test Results')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# two-sample t-test results
plt.hist(group1, bins=5, alpha=0.7, color='blue', edgecolor='black', label='Group 1')
plt.hist(group2, bins=5, alpha=0.7, color='green', edgecolor='black', label='Group 2')
plt.axvline(np.mean(group1), color='red', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(group2), color='orange', linestyle='dashed', linewidth=1)
# overall mean line
overall_mean = np.mean(group1 + group2)
plt.axvline(overall_mean, color='purple', linestyle='dashed', linewidth=1, label='Overall Mean')

plt.title('Two-sample t-test Results')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# paired t-test results
plt.hist(before_treatment, bins=5, alpha=0.7, color='blue', edgecolor='black', label='Before Treatment')
plt.hist(after_treatment, bins=5, alpha=0.7, color='green', edgecolor='black', label='After Treatment')
plt.axvline(np.mean(before_treatment), color='red', linestyle='dashed', linewidth=1)
plt.axvline(np.mean(after_treatment), color='orange', linestyle='dashed', linewidth=1)
# overall mean line
overall_mean = np.mean(before_treatment + after_treatment)
plt.axvline(overall_mean, color='purple', linestyle='dashed', linewidth=1, label='Overall Mean')
plt.title('Paired t-test Results')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
```

```python
# 2.perform a Chi Square test
from scipy.stats import chi2_contingency
# contigency table
data=[[50,30],[20,40]]
# perform Chi-squared test
chi2, p, dof, excepted = chi2_contingency(data)
print("chi-squared statistics: ", chi2)
print("p-value: ",p)
print("excepted frequencies: \n", excepted)

# plotting Chi-Square test results
import matplotlib.pyplot as plt
import numpy as np

# here we use ax to plot observed vs expected frequencies
# ax means axis of the plot
# ax.bar means bar plot
# instead of plt we use ax to customize the plot because ax provides more control over the plot elements

labels = ['Group 1', 'Group 2']
observed = [50, 30]
expected = [35, 45]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, observed, width, label='Observed')
rects2 = ax.bar(x + width/2, expected, width, label='Expected')
ax.set_ylabel('Frequencies')
ax.set_title('Chi-Square Test Results')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
# 3.Conduct ANOVA
from scipy.stats import f_oneway
# data for three groups
group1=[12,14,15,16,17]
group2=[11,13,14,15,16]
group3=[10,12,13,14,15]

# perform ANOVA
f_stat, p_value=f_oneway(group1, group2, group3)    
print("f_statistics:",f_stat)
print("p_values: ",p_value)

# plotting ANOVA results
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data = [group1, group2, group3]
sns.boxplot(data=[group1, group2, group3], palette="Set3")
plt.title('ANOVA Results')
plt.ylabel('Values')
plt.xlabel('Groups')
plt.legend()
plt.show()
```

```python
# 4.perform a two-way ANOVA to test for interaction effects between two independent variables on a dependent variable using scipy and statsmodels

import pandas as pd
import statsmodels.api as sm
from scipy.stats import f_oneway
from statsmodels.formula.api import ols

# create a sample dataset
data = {
    'independent_var1': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'independent_var2': ['X', 'Y', 'Z', 'X', 'Y', 'Z', 'X', 'Y', 'Z'],
    'dependent_var': [5, 6, 7, 8, 9, 10, 11, 12, 13]
}   
df = pd.DataFrame(data)

# perform two-way ANOVA
f_stat, p_value = f_oneway(
    df[df['independent_var1'] == 'A']['dependent_var'],
    df[df['independent_var1'] == 'B']['dependent_var'],
    df[df['independent_var1'] == 'C']['dependent_var']
)
print("Two-way ANOVA results:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
```
