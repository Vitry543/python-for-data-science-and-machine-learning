```python
# Hypothesis tetsitng and P_values

# Hypothesis testing: statistical method used to make decisions about population parameters based on sample data
    # - or statistical method to determine if there is enough evidence in a sample to infer a conclusion about the population parameter 
    # - it involves formulating null and alternative hypotheses and using statistical tests to evaluate the evidence 

    # -Null hypothesis(H0): it represents the status quo or default assumption (assumes no effect or no difference)
    # -Alternative hypothesis(H1): it represents the claim or research hypothesis (assumes an effect or difference)

# Steps in Hypothesis testing:
    # 1.State the null and alternative hypotheses
    # 2.Choose a significance level(alpha)- common values are 0.05, 0.01, 0.001
    # 3.Select an appropriate test statistic 
    # 4.Calculate the test statistic and p-value
    # 5.Make a decision based on the p-value
        # -if p-value < alpha: reject the null hypothesis
        # -if p-value >= alpha: fail to reject the null hypothesis

# P-value: probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true
# -smaller p-value indicates stronger evidence against the null hypothesis

# significance level(alpha): threshold for rejecting the null hypothesis
    # -common values are 0.05, 0.01, 0.001, example: alpha=0.05 means a 5% chance of rejecting the null hypothesis when it is true

# Decision Rule:
    # -Reject the null hypothesis if p-value < alpha
    # -Fail to reject the null hypothesis if p-value >= alpha


# Type of errors:
    # Type I error(alpha): 
    #   -incorrectly rejecting the null hypothesis when it is true
        # -example: rejecting a true null hypothesis - concluding a drug effective when it is not
    # Type II error(beta): 
    #   - failing to reject the null hypothesis when it is false
        # -example: failing to reject a false null hypothesis - concluding a drug not effective when it is

# Examples:

# Reality       Decision                Conclusion
# H0 is true    Reject H0               Type I error
# H0 is true    Fail to reject H0       Correct decision
# H0 is false   Reject H0               Correct decision
# H0 is false   Fail to reject H0       Type II error



# Power of the test: probability of correctly rejecting the null hypothesis when it is false
    # -it is the probability of rejecting the null hypothesis when it is false
    
```

```python
# 1.perform A hypothesis test 

# ttest_1samp: perform a one sample t-test

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_1samp

# sample data
data=[12,14,15,16,17,18,19]

# Null hypothesis: mean=15
population_mean=15

# perform t-test
t_statistic,p_value=ttest_1samp(data,population_mean)

print("t_statistic:",t_statistic)
print("p_value:",p_value)

# interpret results
alpha=0.05
if p_value<alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")


# plot histogram of data
sns.histplot(data,bins=5,kde=True,label="Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Data")
plt.legend()
plt.show()
```

```python
# 2.TWo-Sample t-test

# ttest_ind is used to compare the means of two independent samples

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# sample data
data1=[12,14,15,16,17,18,19]
data2=[10,12,13,14,15,16,17]

# perform t-test
t_statistic,p_value=ttest_ind(data1,data2)

print("t_statistic:",t_statistic)
print("p_value:",p_value)

# interpret results
alpha=0.05
if p_value<alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")

# plot histogram of data
sns.histplot(data1,bins=5,kde=True, label="Data1",color="blue")
sns.histplot(data2,bins=5,kde=True, label="Data2",color="red")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Data")
plt.legend()
plt.show()

sns.histplot(data1,bins=5,kde=True, label="Data1",color="blue")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Data")
plt.legend()
plt.show()
sns.histplot(data2,bins=5,kde=True, label="Data2",color="red")
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Data")
plt.legend()
plt.show()

# plot boxplot of data
sns.boxplot(data=[data1,data2])
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Boxplot of Data")
plt.show()
```

```python
# 3.perform a Z-test for large sample size
url="https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.weightstats import ztest

housing=pd.read_csv(url)

print(housing)

# perform z-test on median_income
z_statistic,p_value=ztest(housing['median_income'],value=26000)

print("z_statistic:",z_statistic)
print("p_value:",p_value)

# plot histogram of median_income
sns.histplot(housing['median_income'],bins=5,kde=True)
plt.xlabel("Median Income")
plt.ylabel("Density of Income")
plt.title("Histogram of Median Income")
plt.grid(True)
plt.show()
```

```python
# use the iris dataset to test if the mean sepal lenght difference between two species

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

url='https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
# load the iris dataset
df=pd.read_csv(url)

# extract the sepal length for the two species
species1=df[df["species"]=="setosa"]["sepal_length"]
species2=df[df["species"]=="versicolor"]["sepal_length"]

# perform t-test
t_statistic,p_value=stats.ttest_ind(species1,species2)

print("t_statistic of species1 and species2 of sepal length:",t_statistic)
print("p_value of species1 and species2 of sepal length:",p_value)

# interpret results
alpha=0.05
if p_value<alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")


# plot histogram of sepal length fo r the two species
plt.hist(species1,bins=5,density=True,label="setosa")
plt.hist(species2,bins=5,density=True,label="versicolor")
plt.xlabel("Sepal Length")
plt.ylabel("Density")
plt.title("Histogram of Sepal Length for the Two Species")

plt.legend()
plt.show()

sns.histplot(species1,bins=5,kde=True,label="setosa")
sns.histplot(species2,bins=5,kde=True,label="versicolor")
plt.xlabel("Sepal Length")
plt.ylabel("Density of Sepal Length")
plt.title("Histogram of Sepal Length for the Two Species")

plt.legend()
plt.show()

```
