````python
# Basic probability Concepts

# 1.Sample space and events
#   -sample space: the set of all possible outcomes of a random Experiment
#   -Events: A subset of the Sample space

#Probability of an event
#   -Probability of an event is the ratio of the number of favorable outcomes to the total number of possible outcomes.

# 2.Types of events
#   -Mutually exclusive events: Events that cannot occur at the same time.
#   -Independent events: Events that do not affect each other.

# 3. Rules of probability
#   -Addition rule: The probability of the union of two events is the sum of their individual probabilities minus the probability of their intersection.
#   -Multiplication rule: The probability of the intersection of two events is the product of their individual probabilities.

# 4. Conditional probability
#   -Conditional probability is the probability of an event given that another event has occurred.
#   -Conditional probability is calculated as the probability of the intersection of two events divided by the probability of the first event.

# 5. Bayes' theorem
#   -Bayes' theorem is a formula for calculating the probability of an event given that another event has occurred.
#   -Bayes' theorem is calculated as the probability of the intersection of two events divided by the probability of the first event.

# 6.independent events
#   -Independent events are events that do not affect each other.
#   -The probability of the intersection of two independent events is the product of their individual probabilities.
#   -P(A ∩ B) = P(A) * P(B)

# 7. Dependent events
#   -Dependent events are events that affect each other.
#   -The probability of the intersection of two dependent events is the product of their individual probabilities.
#   -P(A ∩ B) = P(A) * P(B|A)

# 8. Complementary events
#   -Complementary events are events that are opposite of each other.
#   -The probability of the intersection of two complementary events is the product of their individual probabilities.
#   -P(A) + P(A^c) = 1

# 9. Mutually exclusive events
#   -Mutually exclusive events are events that cannot occur at the same time.
#   -The probability of the intersection of two mutually exclusive events is the product of their individual probabilities.
#   -P(A U B) = P(A) + P(B)

# 10. Mutually independent events
#   -Mutually independent events are events that do not affect each other.
#   -The probability of the intersection of two mutually independent events is the product of their individual probabilities.
#   -P(A U B) = P(A) + P(B) - P(A ∩ B)
````

````python
# Random Variables
# -random variables are functions that map outcomes of a random experiment to real numbers

# types of random variables
# 1. Discrete random variables: it takes on countable values
# 2. Continuous random variables: it takes on uncountable values


# Probability Mass Function (PMF)
# -PMF is a function that gives the probability of a discrete random variable taking on a specific value
# example: rolling a die, with outcomes 1,2,3,4,5,6


# Probability Density Function (PDF)
# -PDF is a function that gives the probability of a continuous random variable taking on a specific value
# example: height of a person
````

````python
# Expectation, Variance, and Standard Deviation

# 1.Expectation: weighted average of all possible values of random variable
# -E(X): weighted average of all possible values of X
# E(X) = summation(x_i * P(X_i=x_i)) for all possible values of X, for discrete random variable
# E(X) = integration(x * f(x)) for all possible values of X, for continuous random variable

# where X is random variable
# x_i is possible value of X
# P(X_i=x_i) is probability of X_i
# f(x) is probability density function of X


# 2.Variance: measure of spread of random variable
# -Var(X): measure of spread of X
# Var(X) = E[(X-E(X))^2]
# Var(X) = summation((x_i-E(X))^2 * P(X_i=x_i)) for all possible values of X, for discrete random variable
# Var(X) = integration((x-E(X))^2 * f(x)) for all possible values of X, for continuous random variable

# 3.Standard Deviation(sigma of x): square root of variance
# -SD(X): square root of variance
# SD(X) = sqrt(Var(X))
# SD(X) = sqrt(E((X-E(X))^2))
# SD(X) = sqrt(summation((x_i-E(X))^2 * P(X_i=x_i))) for all possible values of X, for discrete random variable
# SD(X) = sqrt(integration((x-E(X))^2 * f(x))) for all possible values of X, for continuous random variable
````

````python
import numpy as np

# random Variables: dice roll

outcomes= np.array([1,2,3,4,5,6])
probabilities= np.array([1/6]*6)

# np.random.choice(outcomes, p=probabilities)

# expectation
expectation= np.sum(outcomes*probabilities)
print("expectation (mean):",expectation)

# variance
variance= np.sum((outcomes-expectation)**2*probabilities)
print("variance:",variance)

# standard deviation
std_dev= np.sqrt(variance)
print("standard deviation:",std_dev)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulating 10000 dice rolls
rolls = np.random.randint(1, 7, size=10000)

# Calculate probabilities
p_even = np.sum(rolls % 2 == 0) / len(rolls)
p_greater_than_4 = np.sum(rolls > 4) / len(rolls)

print("Probability of getting even numbers:", p_even)
print("Probability of getting greater than 4:", p_greater_than_4)

# Count occurrences of each face
faces = [1, 2, 3, 4, 5, 6]
counts = [np.sum(rolls == i) for i in faces]

# Create color-coded bar chart
colors = []
for face in faces:
    if face % 2 == 0 and face > 4:
        colors.append('purple')  # Both conditions
    elif face % 2 == 0:
        colors.append('blue')    # Even only
    elif face > 4:
        colors.append('red')     # Greater than 4 only
    else:
        colors.append('gray')    # Neither

plt.figure(figsize=(10, 6))
plt.bar(faces, counts, color=colors, edgecolor='black')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Dice Roll Distribution\nBlue=Even, Red=>4, Purple=Both, Gray=Neither')
plt.xticks(faces)

# Add probability annotations
plt.text(6.5, max(counts)*0.9, f'P(Even) = {p_even:.4f}', fontsize=10)
plt.text(6.5, max(counts)*0.8, f'P(>4) = {p_greater_than_4:.4f}', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
# Here's a simple and clear solution using only basic plotting functions you know (`plt.bar`, `plt.hist`):


import numpy as np
import matplotlib.pyplot as plt

# Simulating 10000 dice rolls
rolls = np.random.randint(1, 7, size=10000)

# Calculate the probability of getting even numbers
p_even = np.sum(rolls % 2 == 0) / len(rolls)
p_greater_than_4 = np.sum(rolls > 4) / len(rolls)

print("Probability of getting even numbers:", p_even)
print("Probability of getting greater than 4:", p_greater_than_4)

# Create two subplots side by side
plt.figure(figsize=(12, 5))

# Plot 1: Histogram of all dice rolls
plt.subplot(1, 2, 1)
plt.hist(rolls, bins=6, range=(0.5, 6.5), color='lightblue', edgecolor='black')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Dice Roll Distribution')

# Plot 2: Bar chart showing the two probabilities
plt.subplot(1, 2, 2)
categories = ['Even Numbers', 'Greater than 4']
probabilities = [p_even, p_greater_than_4]
plt.bar(categories, probabilities, color=['blue', 'red'])
plt.ylabel('Probability')
plt.title('Calculated Probabilities')
plt.ylim(0, 1)  # Set y-axis from 0 to 1 since probabilities are between 0 and 1

# Add the probability values on top of each bar
for i, prob in enumerate(probabilities):
    plt.text(i, prob + 0.02, f'{prob:.3f}', ha='center')

plt.tight_layout()
plt.show()

# This code uses only:
# - `plt.hist()` to show your dice roll distribution
# - `plt.bar()` to show your calculated probabilities (`p_even` and `p_greater_than_4`)
# - Basic functions like `plt.subplot()`, `plt.xlabel()`, `plt.ylabel()`, `plt.title()`

# The left plot shows how many times each dice face appeared in your 10,000 rolls.  
# The right plot shows your two calculated probabilities as bars with their exact values displayed on top.
````

````python
# Here's a beginner-friendly solution using only basic matplotlib functions (bar, scatter, plot, hist) that you're familiar with:

# ```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate dice rolls
rolls = np.random.randint(1, 7, size=10000)

# Calculate probabilities
p_even = np.mean(rolls % 2 == 0)
p_greater_than_4 = np.mean(rolls > 4)

print(f"Probability of even numbers: {p_even:.4f}")
print(f"Probability of >4: {p_greater_than_4:.4f}")

# Create a figure with 3 subplots
plt.figure(figsize=(15, 4))

# 1. Plot all rolls (histogram)
plt.subplot(1, 3, 1)
plt.hist(rolls, bins=np.arange(0.5, 7.5, 1), edgecolor='black')
plt.title('All Dice Rolls')
plt.xlabel('Dice Value')
plt.ylabel('Count')
plt.xticks(range(1, 7))

# 2. Plot even numbers (bar chart)
plt.subplot(1, 3, 2)
even_counts = np.array([np.sum(rolls == i) for i in range(1, 7)])
colors = ['blue' if i%2 == 0 else 'lightgray' for i in range(1, 7)]
plt.bar(range(1, 7), even_counts, color=colors, edgecolor='black')
plt.title(f'Even Numbers (P={p_even:.3f})')
plt.xlabel('Dice Value')
plt.ylabel('Count')
plt.xticks(range(1, 7))

# 3. Plot numbers >4 (bar chart)
plt.subplot(1, 3, 3)
gt4_counts = np.array([np.sum(rolls == i) if i > 4 else 0 for i in range(1, 7)])
colors = ['red' if i > 4 else 'lightgray' for i in range(1, 7)]
plt.bar(range(1, 7), gt4_counts, color=colors, edgecolor='black')
plt.title(f'Numbers >4 (P={p_greater_than_4:.3f})')
plt.xlabel('Dice Value')
plt.ylabel('Count')
plt.xticks(range(1, 7))

plt.tight_layout()
plt.show()
# # ```

# This solution:
# 1. Uses only basic plotting functions you know (hist, bar)
# 2. Shows three clear visualizations side-by-side
# 3. Colors the relevant bars (blue for even, red for >4)
# 4. Includes probability values in the titles
# 5. Has proper labels and ticks for easy understanding

# The first plot shows all rolls, the second highlights even numbers, and the third highlights numbers greater than 4. This makes it easy to compare the probabilities visually.
````
