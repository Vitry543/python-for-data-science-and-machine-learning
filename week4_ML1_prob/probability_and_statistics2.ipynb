```python
# Common probability Distributions:

# 1.Gaussian(Normal) Distribution:
    # -Bell-Shaped Curve characterized by mean(mu) and standard devitaion(sigma)
    # -it follows probability density function: pdf is f(x)=(1/(2*pi*sigma^2)**0.5)*e^(-((x-mu)**0.5)/2*sigma**2)

    # properties:
    # -symmetric about the mean
    # -mean, median and mode are the same

    # Application in ML:
    # -common assumption in many algorithms(eg: Naive bayes)
    # -Used in feature scaling(eg: Standardization)

# 2.Binomal Distribution:
    # -Models the number of successes in n, n indepedent Bernoulli trial
    # -it follows probability mass function(PMF)
    # -P(X=k)=c(n k)p^k * (1-p)^(n-k)

    # properties
    # -Discrete distribution
    # -parameter: n(no. of trials), p(probability of success)

    #Application in ML:
    # -Logistic regression assumes a binomial distribution for binary classification

# 3.possion distribution:
    # -Models the number of events in a fixed interval
    # follows pmf
    # pmf: P(X=k)=((λ^k) * e^-λ)/k!

    # properties:
    # - Discrete distribution
    # -parameter: λ (average rate of occurrence)

    # Application in ML:
    # -Used in event Modeling
    # -ex: clicks on website

# 4.Uniform Distribution:
    # -All outcomes are equally likely within a specified range [a,b]
    # it follows pdf:
    # pdf: f(x)=1/(b-a) for a<=x<=b else 0

    # properties:
    # -Continuous distribution
    # -parameter: a(lower bound), b(upper bound)

    # Application in ML:
    # -Used in random sampling
    # -Initialization of weights in neural networks

# 5.Exponential Distribution:
    # -Models the time between events in a Poisson process
    # -it follows pdf:  
    # pdf: f(x;λ)=λ * e^(-λx) for x>=0 else 0

    # properties:
    # -Continuous distribution
    # -parameter: λ(rate parameter)
    # -memoryless property

    # Application in ML:
    # -Used in survival analysis
    # -Modeling time until an event occurs

# 6.Bernoulli Distribution:
    # -Models a single trial with two possible outcomes(success or failure)
    # -it follows pmf:
    # pmf: P(X=1)=p and P(X=0)=1-p

    # properties:
    # -Discrete distribution
    # -parameter: p(probability of success)

    # Application in ML:
    # -Used in binary classification tasks
    # -Building block for more complex models(like binomial distribution)

# 7.Categorical Distribution:
    # -Generalization of Bernoulli distribution for multi-class outcomes
    # -it follows pmf:
    # pmf: P(X=i)=p_i for i=1,2,...,k where sum of p_i=1

    # properties:
    # -Discrete distribution
    # -parameter: p_1, p_2,...,p_k (probabilities of each category)


    # Application in ML:    # -Used in multi-class classification tasks
    # -Modeling outcomes with more than two categories  

# 8.Multinomial Distribution:
    # -Generalization of binomial distribution for multi-class outcomes
    # -it follows pmf:
    # pmf: P(X_1=k_1, X_2=k_2,...,X_k=k_k)=(n!/(k_1! * k_2! *...* k_k!)) * p_1^k_1 * p_2^k_2 *...* p_k^k_k 

    # properties:
    # -Discrete distribution
    # -parameter: n(total trials), p_1, p_2,...,p_k (probabilities of each category)

    # Application in ML:
    # -Used in text classification tasks(eg: Naive bayes for text)
    # -Modeling outcomes with multiple categories over several trials
    # -Used in multi-class classification tasks    
    # # -Modeling outcomes with more than two categories


```

```python
# Applications of Probability Distributions in Machine Learning:
    # -Data Modeling: Understanding data distribution for better model selection
    # -Feature Engineering: Transforming features based on their distributions  
    # -Uncertainty Estimation: Quantifying uncertainty in predictions
    # -Anomaly Detection: Identifying outliers based on probability thresholds
#  --- IGNORE ---

# applications_of_Distributions_in_Machine_Learning:

# 1.Gaussian Distribution:
    # -Used in algorithms like Gaussian Naive Bayes and Gaussian Mixture Models
    # it is assumed in many statistical models due to its natural occurrence in real-world data

# 2.Binomial Distribution:
    # -Used in Logistic Regression for binary classification tasks
    # -Helps in modeling the number of successes in a fixed number of trials

# 3.Poisson Distribution:
    # -Used in modeling count data, such as the number of events occurring within a fixed interval
    # -Applicable in scenarios like predicting the number of customer arrivals or website clicks

# 4.Uniform Distribution:
    # -Used in random sampling techniques and initializing weights in neural networks
    # -Helps in scenarios where all outcomes are equally likely

```

```python
# Visualizing Distributions and understand their properties:

# visualization helps understand the skewness(), kurtosis(), and outliers()
# 1.Skewness:
    # -measures of symmetry: positive skew(right tail longer), negative skew(left tail longer), zero skew(symmetric)
    # -calculated using formula: skewness= (n/((n-1)(n-2))) * Σ((xi - x̄)/s)^3

# 2.Kurtosis:
    # -measures the "tailedness" of the distribution: high kurtosis
    # -it has high kurtosis: heavy tails and a sharp peak, while low kurtosis has light tails and a flat peak

# 3.outliers:
    # -data points that deviate significantly from the rest of the data
    # -can be detected using statistical methods(z-score, IQR) or visualization techniques(box plots, scatter plots)
# -outliers can impact model performance and should be handled appropriately(detection, treatment)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, uniform
import seaborn as sns

# guassian distribution visualization
mu, sigma = 0, 0.1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, label="gaussian")
plt.title('Gaussian Distribution (mu=0, sigma=0.1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.show()

# or 
x=np.linspace(-5,5,1000)
plt.plot(x,norm.pdf(x,loc=0, scale=1), label="guassian")
plt.title('Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```

```python
# binomal distribution

n,p=10, 0.5
x=np.arange(0,11)
y=binom.pmf(x,n,p)
plt.bar(x,y, alpha=0.5, label='binomial dist(n=10,p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.grid(True)
plt.legend()
plt.show()

# poisson distribution
lam=3
x=np.arange(0,11)
y=poisson.pmf(x,lam)
plt.bar(x,y, alpha=0.5, label='poisson dist(lam=3)', color='green')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.grid(True)
plt.legend()
plt.show()

# uniform distribution
a,b=0,10
x=np.arange(a,b+1)
y=uniform.pdf(x,a,b)
plt.bar(x,y, alpha=0.5, label='uniform dist(a=0,b=10)', color='blue')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Uniform Distribution')
plt.grid(True)
plt.legend()
plt.show()

# or 
x=np.random.uniform(low=0,high=10,size=1000)
y=uniform.pdf(x,a,b)
sns.histplot(x,bins=30,kde=True, label='uniform dist(a=0,b=10)', color='red')
plt.show()

# or 
x=np.random.uniform(low=0,high=10,size=1000)
y=uniform.pdf(x,a,b)
plt.hist(x,bins=30, density=True, alpha=0.5, label='uniform dist(a=0,b=10)', color='red')
plt.show()

# the above uniform distribution is a continuous distribution and all three are same,\
#  but the first one is a discrete distribution, and the second one is a continuous distribution , and the third one is a histogram of the uniform distribution

```

```python
# 2.analyze a Dataset's Distribution
# example; load skewness and kurtosis load


url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'

from scipy.stats import skew, kurtosis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(url)

# analyze the sepal_length
feature=df['sepal_length']
print("skewness: ",skew(feature))
print("kurtosis: ",kurtosis(feature))

# visualize the distribution
sns.histplot(feature,kde=True, bins=20, label='Sepal Length')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
