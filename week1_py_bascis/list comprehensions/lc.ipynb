```python
# Normal form
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def find_primes_up_to(limit):
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

limit = 50
print(f"Prime numbers up to {limit}: {find_primes_up_to(limit)}")

# List comprehension form
def find_primes_up_to(limit):
    return [num for num in range(2, limit + 1) if is_prime(num)]

limit = 50
print(f"Prime numbers up to {limit}: {find_primes_up_to(limit)}")
```

```python

# List comprehension form
limit = 50
primes = [n for n in range(2, limit + 1) if 
          all(n % i != 0 for i in range(2, int(n**0.5) + 1))]
print(f"Prime numbers up to {limit} (list comprehension): {primes}")
```
