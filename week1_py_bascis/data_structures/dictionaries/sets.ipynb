```python
# adding and removing elements
numbers.add(11)
print(numbers)
numbers.remove(11)
print(numbers)
numbers.discard(10)
print(numbers)
numbers.pop()
print(numbers)
```

```python
# set operations
s={1,2,3,4,5}
t={4,5,6,7,8}
print(s.union(t))
print(s.intersection(t))
print(s.difference(t))
print(s.symmetric_difference(t))
```

```python
s={1,2,3,4,5}
t={4,5,6,7,8}
print(s&t)#intersection: common elements
print(s-t)#difference: elements in s but not in t
print(t-s)#difference: elements in t but not in s
print(s|t)#union: all elements in both sets
print(s^t)#symmetric difference: elements in s or t but not in both
```
