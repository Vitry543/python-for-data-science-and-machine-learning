```python
import os

print(os.getcwd())
if not os.path.exists('test_dir.txt'):
    os.mkdir('test_dir.txt') #here mkdir is used to create the directory
else:
    os.rmdir('test_dir.txt') #here rmdir is used to remove the directory

```

```python
if os.path.exists('file.txt'):
    os.remove('file.txt')
else:
    os.mkdir('file.txt')
```
