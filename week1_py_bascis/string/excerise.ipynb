```python
# create a text cleaner
import re
def clean_text(text):
    text = re.sub(r"[^a-zA-Z0-9 ]", "", text) #here ^ is used to match any character that is not in the set
    # AND ^a-zA-Z0-9 is used to match any character that is not in the set. HERE a-z, A-Z, 0-9 are in the set
    text = re.sub(r"\s+", " ", text)
    return text.lower()

print(clean_text("Hello World!"))
```

```python
import re
def clean_text(text):
    # remove punctuation
    text=re.sub(r'[^\w\s]','',text) #here \w matches any word character and \s matches any whitespace character
# word character means the characters that are part of a word
    # remove extra spaces
    # text=' '.join(text.split()) #here split() splits the string into a list of words and '' joins the list of words into a string
    text=re.sub(r'\s+',' ',text) #here \s+ matches one or more whitespace characters
# whitespace means the space between the words

    # convert to lowercase
    text=text.lower()
    return text

input_text="    Hello,     World.!!!! welcome to Python, programming....."
print(clean_text(input_text))
clean_text(input_text)
```

```python
def palindrome(s):
    s="".join(char.lower() for char in s if char.isalnum())
    return s==s[::-1]

print(palindrome("A man, a plan, a canal: Panama"))
```

```python
def palindrome(s):
    for char in s:
        if char.isalpha(): #isaplha 
            s="".join(char.lower())
    return s==s[::-1]

text=input("enter the string: ")
if palindrome(text):
    print(f"{text} is a palindrome")
else:
    print(f"{text} is not a palindrome")
```

```python
# count the number of vowels in a string

def vowels(S):
    count=0
    for char in S:
        if char in "aeiouAEIOU":
            count+=1
    return count

text=input("enter the string: ")
print(vowels(text))

def vowel(s):
    s="".join(char for char in s if char in "aeiouAEIOU")
    return len(s)

print(vowel(text))
```

```python
# program to replace all email address in a text with xxxx using regex
import re
text="hello my email is abc@gmail.com and my another email is def@gmail.com"    
updated_text=re.sub(r"\w+@\w+\.\w+","xxxx",text)
# \w+@\w+\.\w+ here \w+ means one or more word characters, @ means @, \w+ means one or more word characters, \. means ., \w+ means one or more word characters are all the characters in the email address replaced with xxxx
print(updated_text)
```
