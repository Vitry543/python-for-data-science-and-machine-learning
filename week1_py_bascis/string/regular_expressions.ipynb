```python
text="contact me at 123-456-7890"
digits=re.findall(r"\d",text)
print(digits)
digits=re.findall(r"\d+",text)
print(digits)
```

```python
update_text=re.sub(r'\d',"X",text)
print(update_text)

update_text=re.subn(r'\D',"X",text)
print(update_text)
```

```python
# re.match
text="hello world"
print(re.match("hello", text))

text="contact me at 123-456-7890"
print(re.match("123", text))

# re.search
text="contact me at 123-456-7890"
print(re.search("123", text))

# re.findall
text="contact me at 123-456-7890"
print(re.findall("123", text))

# re.finditer : it returns an iterator
text="contact me at 123-456-7890"
print(re.finditer("123", text))

# re.subn: it returns a tuple
text="contact me at 123-456-7890"
print(re.subn("123", "456", text))

# re.sub: it returns a string
text="contact me at 123-456-7890"
print(re.sub("123", "456", text))

# re.split
text="contact me at 123-456-7890"
print(re.split("-", text))
```

```python
# re.compile : it returns a pattern object
# here complie is used to compile the pattern into a regular expression object
text="contact me at 123-456-7890"
pattern=re.compile("123")
print(pattern)
print(pattern.search(text))
print(pattern.match(text)) #here match will check from the start of the string and output will be none if it is not matching at the start
print(pattern.findall(text)) #here findall will return all the matches in the string
print(pattern.finditer(text)) #here finditer will return all the matches in the string as an iterator. here iterator is an object that can be iterated over
print(pattern.sub("",text)) #here sub will replace all the matches in the string
print(pattern.subn("",text)) #here subn will replace all the matches in the string and return the number of replacements
```

```python
# re.escape : it returns a string with all the special characters escaped
text="hello world"
print(re.escape(text))

text="contact me at 123-456-7890"
print(re.escape(text))
```
