```python
# program to copy the content of one file to another file
def read_the_file(filename):
    try:
        with open(filename,'r') as file:
            content=file.read()
            print(content)
    except FileNotFoundError:
        print("file not found")

read_the_file("sample.txt")

def copy_content(source_file, destination_file):
    try:
        with open(source_file, 'r') as source:
            content = source.read()
        with open(destination_file, 'w') as destination:
            destination.write(content)
        print("Content copied successfully.")
    except FileNotFoundError:
        print("Source file not found.")

copy_content("sample.txt", "sample1.txt")

with open('sample1.txt','r') as file:
    print(file.read())
```

```python
# program that counts the number of a occurences of a specific word in a text file

def count_word_occurrences(filename, target_word):
    """
    Counts how many times target_word appears in the file specified by filename.
    """
    try:
        with open(filename, 'r') as file:
            # Read the file content
            text = file.read()
            
            # Split text into words
            words = text.split()
            
            # Count occurrences (case-insensitive)
            count = 0
            target_word = target_word.lower()
            
            for word in words:
                # Remove punctuation for cleaner matching and convert to lower case
                clean_word = word.strip('.,!?;:"()[]{}').lower()
                if clean_word == target_word:
                    count += 1
            
            return count
            
    except FileNotFoundError:
        return f"Error: The file '{filename}' was not found."

if __name__ == "__main__":
    # Create a temporary file for testing
    test_file = "sample.txt"
    with open(test_file, "w") as f:
        f.write("Hello world. Hello Python. Coding is fun. Say hello!")
    
    word_to_find = "hello"
    result = count_word_occurrences(test_file, word_to_find)
    print(f"The word '{word_to_find}' appears {result} times in '{test_file}'.")
```

```python
# The name of the file to be read
file_name = "sample.txt"

try:
    # 1. Ask the user for the word they want to count.
    word_to_count = input("Enter the word you want to search for: ")

    # 2. Open and read the file.
    # The 'with' statement ensures the file is properly closed.
    with open(file_name, 'r') as file:
        content = file.read()

    # 3. Prepare the text for accurate counting.
    #    - Convert all text to lowercase to make the search case-insensitive.
    #    - Replace common punctuation with a space to isolate words.
    text_to_process = content.lower()
    punctuation = '.,!?;:"()[]'
    for char in punctuation:
        text_to_process = text_to_process.replace(char, ' ')

    # 4. Split the processed text into a list of individual words.
    word_list = text_to_process.split()
    
    # 5. Count the occurrences of the user's word (also converted to lowercase).
    occurrence_count = word_list.count(word_to_count.lower())

    # 6. Display the final result to the user.
    print(f"\nThe word '{word_to_count}' appears {occurrence_count} times in the file.")

except FileNotFoundError:
    print(f"\nError: The file '{file_name}' was not found.")
    print("Please make sure a file named 'sample.txt' exists in the same directory as this script.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# write a program to log messages with timestamps into a file

from datetime import datetime

def log_message(filename, message):
    """
    Appends a message with the current timestamp to the specified file.
    """
    # Get current date and time formatted as Year-Month-Day Hour:Minute:Second
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Create the formatted log entry
    entry = f"[{current_time}] {message}\n"
    
    # Open file in append mode ('a') so we don't overwrite existing logs
    with open(filename, 'a') as file:
        file.write(entry)

if __name__ == "__main__":
    log_file = "system_logs.txt"
    log_message(log_file, "Process started.")
    log_message(log_file, "Data loaded successfully.")
    log_message(log_file, "Process finished.")
    print(f"Logs written to {log_file}")
```
