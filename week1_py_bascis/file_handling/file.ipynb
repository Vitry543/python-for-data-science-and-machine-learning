```python
with open('text.txt','w') as file:
    file.write("Hello World")

with open('text.txt','r') as file:
    print(file.read())
```

```python
# opening a file can be done with open() and r|w|a|r+
# r: read,w:write, a:append, r+: write and read

# reading files
# read(), readline(), readlines()

# writing files
# write(), writelines()
    
```

```python
with open('sample.txt','w') as file:
    file.write('Hello World')
    file.writelines(['alice', 'boo','john'])
```

```python
with open('sample.txt','r+') as file:
    print(file.read())
    print(file.readline())
    file.write("\ngood morning")
    print(file.read())
```

```python
with open("sample.txt",'a') as file:
    file.write("\ngood morning")
    
```

```python
try:
    with open('sample.txt','r') as file:
        content=file.read()
        print(content)
except FileNotFoundError:
    print("file not found")
```
